**Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.**

`Matplotlib` is a popular Python library used for creating various types of plots and visualizations. It is commonly used for data visualization tasks in fields like data analysis, scientific research, and more.

The `Pyplot` module in Matplotlib provides a simple and convenient interface for creating a wide range of plots. Here are five types of plots that can be plotted using the Pyplot module of Matplotlib:

- `Line Plot:` Visualizes data points connected by straight lines, commonly used for time series data.
- `Scatter Plot:` Displays individual data points as markers, useful for showing the relationship between two variables.
- `Bar Plot:` Represents categorical data with rectangular bars, suitable for comparing quantities across categories.
- `Histogram:` Divides data into intervals and shows the frequency or count of data points in each interval.
- `Pie Chart:` Represents data as slices of a circular pie, showing the proportion of each category relative to the whole.

`Now take a look at the examples of these plots:`

````python
import matplotlib.pyplot as plt  # Import the Matplotlib library

# Create sample data
x = [1, 2, 3, 4, 5]  # X-axis values
y = [10, 25, 15, 30, 20]  # Y-axis values

# Line plot
plt.figure(figsize=(8, 6))  # Create a new figure
plt.plot(x, y, marker='o')  # Plot the data with markers at data points
plt.title("Line Plot")  # Add a title to the plot
plt.xlabel("X-axis")  # Label for the X-axis
plt.ylabel("Y-axis")  # Label for the Y-axis
plt.grid(True)  # Add grid lines
plt.show()  # Display the plot
````

**Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.**

```python
import numpy as np

np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)

y = 3 + np.random.normal(0, 2, len(x))
```

`Note: Also add title, xlabel, and ylabel to the plot.`

A `scatter plot` is a type of data visualization that displays individual data points as dots on a `two-dimensional` plane. It is used to show the relationship between two continuous variables. Here's the completed code for generating and plotting a scatter plot using the given data:

````python
import matplotlib.pyplot as plt
import numpy as np

# Generate data
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

# Scatter plot
plt.figure(figsize=(8, 6))
# Plot the data with square markers
plt.scatter(x, y, color='red', marker='s')  
plt.title("Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()
````

**Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.**

**Use the following data:**

`import numpy as np`

`For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])`

`For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])`

`For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])`

`For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])`

The `subplot()` function is used to create a `grid of subplots` within a `single` figure. It allows you to display `multiple plots` in a single window, making it easier to compare and analyze different datasets or visualizations.

````python
import matplotlib.pyplot as plt
import numpy as np

# Data for the four lines
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create a 2x2 grid of subplots
plt.figure(figsize=(12, 10))

# Subplot 1: Line plot 1
plt.subplot(2, 2, 1)
plt.plot(x1, y1, marker='o', color='blue')
plt.title("Line Plot 1")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Subplot 2: Line plot 2
plt.subplot(2, 2, 2)
plt.plot(x2, y2, marker='s', color='green')
plt.title("Line Plot 2")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Subplot 3: Line plot 3
plt.subplot(2, 2, 3)
plt.plot(x3, y3, marker='^', color='red')
plt.title("Line Plot 3")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Subplot 4: Line plot 4
plt.subplot(2, 2, 4)
plt.plot(x4, y4, marker='d', color='purple')
plt.title("Line Plot 4")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Adjust layout to prevent overlapping
plt.tight_layout()

# Display the subplots
plt.show()
````

**Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.**

```python
import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])
```

A `bar plot` is a graphical representation of data using `rectangular bars.` Each bar's `length` or `height` is proportional to the value it represents. `Bar plots` are commonly used to display and compare the values of different categories or groups.

In our case, we can use the `matplotlib` library to create both a `bar plot` and a `horizontal bar plot` for the given data. Here's the code to do that:

````python
import matplotlib.pyplot as plt
import numpy as np

# Generate Data
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Create a bar plot
plt.figure(figsize=(8, 6))
plt.bar(company, profit, color='blue')
plt.title("Company Profits")
plt.xlabel("Company")
plt.ylabel("Profit")
plt.show()

# Create a horizontal bar plot
plt.figure(figsize=(8, 6))
plt.barh(company, profit, color='green')
plt.title("Company Profits")
plt.xlabel("Profit")
plt.ylabel("Company")
plt.show()
````

**Q5: What is a box plot? Why is it used? Using the following data plot a box plot.**

`box1 = np.random.normal(100, 10, 200)`

`box2 = np.random.normal(90, 20, 200)`

A `box plot`, also known as a `box-and-whisker plot`, is a graphical representation of the distribution of a dataset. It displays the `median`, `quartiles`, and `potential outliers` of the data. It's used to summarize and visualize the `spread`, `skewness`, and `central tendency` of the data, making it easier to compare multiple datasets or identify any unusual values.

Here's the code to create box plots for the given data using the `matplotlib library:`

````python
import matplotlib.pyplot as plt
import numpy as np

# Generate Data
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Create a box plot
plt.figure(figsize=(8, 6))
plt.boxplot([box1, box2], labels=['Box 1', 'Box 2'])
plt.title("Box Plot")
plt.xlabel("Box")
plt.ylabel("Value")
plt.show()
````

- `Median:` The line inside the box represents the median, which is the `middle` value of the dataset.
- `Quartiles:` The box itself represents the `interquartile range (IQR)`, the range between the `25th` and `75th` percentiles.
- `Whiskers:` The lines extending from the box `(whiskers)` typically show `1.5` times the IQR.
- `Outliers:` Points beyond the `whiskers` are considered `potential outliers` and are shown as individual points.

- `Spread:` The length of the box (IQR) indicates how spread out the `middle 50%` of the data is.
- `Central Tendency:` The `median` represents the central tendency of the data.
- `Skewness:` The direction and degree of skewness of the data can be inferred based on the relative positions of the median and the whiskers.
  1. If the `median` of a box is closer to one end of the box, it might indicate skewness in that direction.
  2. If one `whisker` is `longer` than the other, it can suggest `skewness` toward the `longer whisker.`

`Code Explanation:`

- This code generates two datasets `box1` and `box2` using numpy's `random.normal` function.
- It then creates a `box plot` using matplotlib's `plt.boxplot` function.
- The labels parameter provides names for the boxes.
- Finally, the plot is given a title and axis labels, and it's displayed using `plt.show().`
