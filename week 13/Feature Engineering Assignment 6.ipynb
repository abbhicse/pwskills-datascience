**Q1. Pearson correlation coefficient is a measure of the linear relationship between two variables. Suppose you have collected data on the amount of time students spend studying for an exam and their final exam scores. Calculate the Pearson correlation coefficient between these two variables and interpret the result.**

`let's go through the process step by step:`

**Step 1: Collect Data**
Let's say we have the following data on the amount of time students spend studying $X$ and their final exam scores $Y$:

$$\begin{align*}
X & : [10, 20, 30, 40, 50] \text{ (hours of study)} \\
Y & : [50, 60, 70, 80, 90] \text{ (exam scores)}
\end{align*}$$

**Step 2: Calculate the Mean**
Calculate the mean $\bar{X}$ and $\bar{Y}$ of the two variables:

$$\begin{align*}
\bar{X} & = \frac{1}{n}\sum_{i=1}^{n}X_i \\
\bar{Y} & = \frac{1}{n}\sum_{i=1}^{n}Y_i
\end{align*}$$

**Step 3: Calculate the Covariance**
Calculate the covariance $S_{XY}$ between $X$ and $Y$:

$$S_{XY} = \frac{\sum_{i=1}^{n}(X_i - \bar{X})(Y_i - \bar{Y})}{n}$$

**Step 4: Calculate the Standard Deviations**
Calculate the standard deviations $S_X$ and $S_Y$:

$$S_X = \sqrt{\frac{\sum_{i=1}^{n}(X_i - \bar{X})^2}{n}}$$

$$S_Y = \sqrt{\frac{\sum_{i=1}^{n}(Y_i - \bar{Y})^2}{n}}$$

**Step 5: Calculate the Pearson Correlation Coefficient**
Finally, calculate the Pearson correlation coefficient $r$:

$$r = \frac{S_{XY}}{S_X \cdot S_Y}$$

`Now, let's calculate this in Python and visualize it using matplotlib:`

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Collect Data
X = np.array([10, 20, 30, 40, 50])
Y = np.array([50, 60, 70, 80, 90])

# Step 2: Calculate the Mean
mean_X = np.mean(X)
mean_Y = np.mean(Y)

# Step 3: Calculate the Covariance
covariance = np.sum((X - mean_X) * (Y - mean_Y)) / len(X)

# Step 4: Calculate the Standard Deviations
std_dev_X = np.sqrt(np.sum((X - mean_X)**2) / len(X))
std_dev_Y = np.sqrt(np.sum((Y - mean_Y)**2) / len(Y))

# Step 5: Calculate the Pearson Correlation Coefficient
r1 = covariance / (std_dev_X * std_dev_Y)
print("The Pearson Correlation Coefficient Without Numpy:", r1)

# Step 2: Calculate the Pearson Correlation Coefficient using numpy library
correlation_matrix = np.corrcoef(X, Y)
print("The Correlation Matrix:")
print(correlation_matrix)
r2 = correlation_matrix[0, 1]
print("The Pearson Correlation Coefficient Via Numpy:", r2)

# Visualization using imshow and colorbar
plt.imshow(correlation_matrix, cmap='coolwarm', vmin=-1, vmax=1)
plt.colorbar(label='Correlation Coefficient (r2)')
plt.xticks([0, 1], ['Hours of Study', 'Exam Scores'])
plt.yticks([0, 1], ['Hours of Study', 'Exam Scores'])
plt.title('Pearson Correlation Coefficient Matrix')
plt.show()
```

- The closer the coefficient is to 1, the stronger the positive linear relationship.
- the closer it is to -1, the stronger the negative linear relationship.
- A coefficient close to 0 indicates a weak or no linear relationship.
- Hence, we can see a stronger positive linear relationship between hours of study of students and thier exam scores.

**Q2. Spearman's rank correlation is a measure of the monotonic relationship between two variables. Suppose you have collected data on the amount of sleep individuals get each night and their overall job satisfaction level on a scale of 1 to 10. Calculate the Spearman's rank correlation between these two
variables and interpret the result.**

A `monotonic relationship` between two variables means that as one variable increases (or decreases), the other variable consistently either increases or decreases. In other words, there is a general trend or pattern in the relationship, and it doesn't necessarily have to be a straight line.

`Example:`

Consider the following dataset:

$$\begin{align*}
\text{X (Hours of Study)} & : 1, 2, 3, 4, 5 \\
\text{Y (Exam Scores)} & : 50, 60, 70, 80, 90 \\
\end{align*}$$

In this example, as the hours of study increase, the exam scores also increase. This is an example of a **monotonic relationship**. Specifically, it's a **positive monotonic relationship** because both variables increase together.

Let's consider an example of a negative monotonic relationship:

`Example:`

Suppose we have the following dataset:

$$\begin{align*}
\text{X (Number of Hours Watching TV)} & : 5, 4, 3, 2, 1 \\
\text{Y (Productivity Level)} & : 20, 30, 40, 50, 60 \\
\end{align*}$$

In this example, as the number of hours watching TV decreases, the productivity level increases. This is an example of a **negative monotonic relationship**. 

`Mathematical Equation:`

Let's use Spearman's rank correlation coefficient $\rho$ to mathematically express the concept of monotonic relationship. For a set of paired observations $(X_i, Y_i)$, the formula for Spearman's rank correlation coefficient is:

$$\rho = 1 - \frac{6 \sum d_i^2}{n(n^2 - 1)}$$

Where:
- $\rho$ is the Spearman's rank correlation coefficient.
- $d_i$ is the difference between the ranks of corresponding values of $X$ and $Y$.
- $n$ is the number of paired observations.

For a **positive monotonic relationship**, most of the $d_i$ values will be positive, indicating that as $X$ increases, $Y$ tends to increase. For a **negative monotonic relationship**, most of the $d_i$ values will be negative, indicating that as $X$ increases, $Y$ tends to decrease.

In our example, if we calculate $\rho$ for the dataset, we would expect a positive value, indicating a positive monotonic relationship.


`Step 1: Understand the Data`

Suppose we have the following data:

$$\begin{align*}
\text{Sleep Hours} & : 7, 5, 8, 6, 7 \\
\text{Job Satisfaction} & : 8, 4, 9, 5, 7 \\
\end{align*}
$$

`Step 2: Rank the Data`

Rank the data for each variable separately. If there are ties, assign the consecutive rank to the tied values.

$$\begin{align*}
\text{Sleep Hours (X)} & : 2, 5, 1, 4, 3 \\
\text{Job Satisfaction (Y)} & : 2, 5, 1, 4, 3 \\
\end{align*}$$

`Step 3: Calculate the Differences and Squares`

Calculate the differences between the ranks for each pair and square the differences.

$$\begin{align*}
\text{Differences (d)} & : 0, 0, 0, 0, 0 \\
\text{Squared Differences (d²)} & : 0, 0, 0, 0, 0 \\
\end{align*}$$

`Step 4: Sum the Squared Differences`

Sum up all the squared differences.

$$\text{Sum of Squared Differences (Σd²)} = 0$$

`Step 5: Calculate Spearman's Rank Correlation Coefficient (\( \rho \))
Use the formula:`

$$\rho = 1 - \frac{6 \times \text{Sum of Squared Differences}}{n \times (n^2 - 1)}$$

- where $n$ is the number of data points.

$$\rho = 1 - \frac{6 \times 0}{5 \times (5^2 - 1)} = 1 - \frac{0}{120} = 1
$$

`Interpretation`

The Spearman's rank correlation coefficient $\rho$ ranges from -1 to 1. 

- $\rho = 1$ indicates a perfect positive monotonic relationship.
- $\rho = -1$ indicates a perfect negative monotonic relationship.
- $\rho = 0$ indicates no monotonic relationship
- $\rho$ very closer to 1 means a strong positive monotonic relationship.
- $\rho$ very closer to -1 means a strong negative monotonic relationship.

In our example, $\rho = 1$, which suggests a perfect positive monotonic relationship between the amount of sleep individuals get each night and their overall job satisfaction level. This means that as the amount of sleep increases, job satisfaction tends to increase as well.

`Python Code with Visualization:`

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import spearmanr

# Data
sleep_hours = np.array([7, 5, 8, 6, 7])
job_satisfaction = np.array([8, 4, 9, 5, 7])

# Calculate ranks
def calculate_ranks(arr):
    # Get the indices that would sort the array in descending order
    sorted_indices = np.argsort(-arr)
    # Initialize an array to store ranks
    ranks = np.zeros_like(arr)
    # Assign ranks to the elements of the original array
    for i in range(len(arr)):
        ranks[sorted_indices[i]] = i + 1

    return ranks


sleep_ranks = calculate_ranks(sleep_hours)
job_satisfaction_ranks = calculate_ranks(job_satisfaction)
# print the values
print("Sleep Hours: ", sleep_hours)
print("Sleep Ranks: ", sleep_ranks)
print("Job Satisfaction: ", job_satisfaction)
print("Job Satisfaction Ranks: ", job_satisfaction_ranks)

# Calculate Spearman's rank correlation coefficient
rho = spearmanr(sleep_hours, job_satisfaction)
# Print Spearman's rank correlation coefficient 
rho_value = rho.statistic
print(f"Spearman's Rank Correlation Coefficient (ρ): {rho_value:.2f}")
```

**Q3. Suppose you are conducting a study to examine the relationship between the number of hours of exercise per week and body mass index (BMI) in a sample of adults. You collected data on both variables for 50 participants. Calculate the Pearson correlation coefficient and the Spearman's rank correlation between these two variables and compare the results.**

**Pearson Correlation Coefficient:**

The Pearson correlation coefficient measures the linear relationship between two continuous variables. It ranges from -1 to 1, where:

- $(r = 1)$ indicates a perfect positive linear relationship,
- $(r = -1)$ indicates a perfect negative linear relationship,
- $(r = 0)$ indicates no linear relationship.

The formula for the Pearson correlation coefficient (\(r\)) is given by:

$$r = \frac{\sum{(X_i - \bar{X})(Y_i - \bar{Y})}}{\sqrt{\sum{(X_i - \bar{X})^2}\sum{(Y_i - \bar{Y})^2}}}$$

Where:
- $X_i$ and $Y_i$ are individual data points.
- $\sum$ denotes the sum
- $\bar{X}$ and $\bar{Y}$ are the means of $X$ and $Y$, respectively.

**Spearman's Rank Correlation:**

Spearman's rank correlation is a non-parametric measure of the monotonicity of the relationship between two variables. It assesses whether there is a consistent increase or decrease in one variable corresponding to a consistent increase or decrease in the other. The formula for Spearman's rank correlation $\rho$ is:

$$\rho = 1 - \frac{6\sum{d_i^2}}{n(n^2 - 1)}$$

Where:
- $d_i$ is the difference between the ranks of corresponding variables.
- $\sum$ denotes the sum
- $n$ is the number of data points.

`Lets give an example with python code:`

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.stats import pearsonr, spearmanr

# Generate example data
np.random.seed(42)
hours_of_exercise = np.random.rand(50) * 10  # 50 random values between 0 and 10
bmi = 25 + 1.5 * hours_of_exercise + np.random.randn(50) * 2  # Simulating a linear relationship with some noise

# Store data in a Pandas DataFrame
data = pd.DataFrame({'Hours_of_Exercise': hours_of_exercise, 'BMI': bmi})
print(data.head())

# Calculate Pearson correlation coefficient
pearson_corr, _ = pearsonr(data['Hours_of_Exercise'], data['BMI'])
# Calculate Spearman's rank correlation
spearman_corr, _ = spearmanr(data['Hours_of_Exercise'], data['BMI'])
# Create a matrix of correlation coefficients for Pearson
pearson_matrix = np.corrcoef(data['Hours_of_Exercise'], data['BMI'])
# print the correlation coefficient values
print(f"Pearson's Correlation Coefficient (ρ): {pearson_corr:.2f}")
print(f"Spearman's Rank Correlation Coefficient (ρ): {spearman_corr:.2f}")

# Visualization using imshow and colorbar
plt.imshow(pearson_matrix, cmap='coolwarm', vmin=-1, vmax=1)
plt.colorbar(label='Correlation Coefficient (r2)')
plt.xticks([0, 1], ['Hours of Study', 'Exam Scores'])
plt.yticks([0, 1], ['Hours of Study', 'Exam Scores'])
plt.title('Pearson Correlation Coefficient Matrix')

plt.show()
```

Interpretation and Comparison:

1. **Pearson's Correlation Coefficient (ρ): 0.92**
   - Interpretation: There is a very strong positive linear relationship between the number of hours of exercise per week and the body mass index (BMI).
   - Strength of Relationship: The value of 0.92 indicates a high positive correlation, suggesting that as the number of hours of exercise per week increases, the BMI tends to increase as well. This positive correlation is close to 1, indicating a strong linear relationship.

2. **Spearman's Rank Correlation Coefficient (ρ): 0.91**
   - Interpretation: There is a very strong monotonic relationship between the number of hours of exercise per week and the body mass index (BMI).
   - Strength of Relationship: The value of 0.91 also indicates a high positive correlation. Spearman's rank correlation is sensitive to the monotonic relationship, not necessarily linear. The similarity between the Pearson and Spearman coefficients suggests that the relationship is not only linear but also monotonic.

**Comparison:**
- In this specific case, both Pearson's and Spearman's correlation coefficients are very similar. This suggests that the relationship between hours of exercise and BMI is not only linear but also monotonic.
- The high positive correlation coefficients indicate a strong tendency for individuals who exercise more to have higher BMIs.

**Q4. A researcher is interested in examining the relationship between the number of hours individuals spend watching television per day and their level of physical activity. The researcher collected data on both variables from a sample of 50 participants. Calculate the Pearson correlation coefficient between these two variables.**

**Calculate the Pearson Correlation Coefficient:**

The Pearson correlation coefficient $r$ is a measure of the strength and direction of a linear relationship between two variables. It ranges from -1 to 1, where:

- $r = 1$ indicates a perfect positive linear relationship,
- $r = -1$ indicates a perfect negative linear relationship, and
- $r = 0$ indicates no linear relationship.

The formula for the Pearson correlation coefficient (\(r\)) is given by:

$$r = \frac{\sum{(X_i - \bar{X})(Y_i - \bar{Y})}}{\sqrt{\sum{(X_i - \bar{X})^2}\sum{(Y_i - \bar{Y})^2}}}$$

Where:
- $X_i$ and $Y_i$ are individual data points.
- $\sum$ denotes the sum
- $\bar{X}$ and $\bar{Y}$ are the means of $X$ and $Y$, respectively.

```python
import pandas as pd

# Sample data
data = {'TV_hours': [3, 4, 2, 5, 1, 6, 4, 7, 3, 5],
        'Physical_activity': [150, 120, 160, 100, 180, 90, 120, 80, 130, 110]}

df = pd.DataFrame(data)
print("The Original DataFrame: ")
print(df)
```

```python
# Calculate the Pearson correlation coefficient
# Calculate means
mean_tv_hours = df['TV_hours'].mean()
mean_phy_activity = df['Physical_activity'].mean()

# Calculate numerator and denominators for the formula
numerator = ((df['TV_hours'] - mean_tv_hours) * (df['Physical_activity'] - mean_phy_activity)).sum()
denom_tv_hours = ((df['TV_hours'] - mean_tv_hours) ** 2).sum()
denom_phy_activity = ((df['Physical_activity'] - mean_phy_activity) ** 2).sum()

# Calculate the Pearson correlation coefficient
r = numerator / (denom_tv_hours * denom_phy_activity)**0.5

print(f"Pearson Correlation Coefficient (r): {r:.2f}")
```

```python
# Visualize the correlation matrix
correlation_matrix = df.corr()
plt.imshow(correlation_matrix, cmap='coolwarm', interpolation='none')
plt.colorbar()
plt.xticks(range(len(correlation_matrix.columns)), correlation_matrix.columns)
plt.yticks(range(len(correlation_matrix.columns)), correlation_matrix.columns)
plt.title('Correlation Matrix')
plt.show()
```

A Pearson correlation coefficient $r$ of -0.98 suggests a very strong negative linear relationship between the number of hours individuals spend watching television per day and their level of physical activity.

Here's how we can interpret this result:

1. **Strength of the Relationship:**
   - The magnitude of the correlation coefficient $r$ is close to 1 (in this case, -0.98), indicating a very strong relationship.

2. **Direction of the Relationship:**
   - The negative sign of the correlation coefficient indicates a negative correlation. In other words, as the number of hours spent watching TV per day increases, the level of physical activity tends to decrease.

3. **Practical Significance:**
   - The strong negative correlation suggests that there is a noticeable and consistent pattern: individuals who spend more time watching TV are likely to have lower levels of physical activity.

4. **Caution:**
   - While correlation coefficients quantify the strength and direction of a linear relationship, they do not imply causation. In this case, a high negative correlation doesn't necessarily mean that watching TV causes a decrease in physical activity or vice versa. Other factors could be influencing this relationship.

**Q5. A survey was conducted to examine the relationship between age and preference for a particular brand of soft drink. The survey results are shown below:**
    
| Age(Years)  |  Soft drink Preference |
|-------------|------------------------|
| 25          |         Coke           |
| 42          |         Pepsi          |
| 37          |       Mountain Dew     |
| 19          |          Coke          |
| 31          |          Pepsi         |
| 28          |          Coke          |

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample data
data = {'Age': [25, 42, 37, 19, 31, 28],
        'Preference': ['Coke', 'Pepsi', 'Mountain Dew', 'Coke', 'Pepsi', 'Coke']}

# Create a DataFrame
df = pd.DataFrame(data)

# Encoding categorical variable 'Preference'
df['Preference'] = df['Preference'].astype('category')
df['Preference'] = df['Preference'].cat.codes

# Calculate the correlation matrix
correlation_matrix = df.corr()

# Extract the correlation coefficient
correlation_coefficient = correlation_matrix.loc['Age', 'Preference']

# Print the correlation coefficient
print(f"Pearson's Correlation Coefficient: {correlation_coefficient:.2f}")

# Visualize the correlation matrix using seaborn heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Pearson Correlation Matrix')
plt.show()
```

The Pearson's correlation coefficient is a measure of the linear relationship between two variables. The value of the coefficient ranges from -1 to 1, where:

- **1:** Perfect positive correlation
- **0:** No correlation
- **-1:** Perfect negative correlation

In our case, the Pearson's correlation coefficient is 0.77. Here's how to interpret it:

1. **Positive Sign (0.77):**
   - The positive sign indicates a positive correlation.
   - As the age increases, there is a tendency for the preference for the soft drink to increase as well.

2. **Strength (0.77):**
   - The closer the coefficient is to 1, the stronger the correlation. In this case, 0.77 indicates a strong positive correlation.

3. **Interpretation:**
   - The Pearson's correlation coefficient of 0.77 suggests a strong positive linear relationship between age and soft drink preference in the provided dataset. This means that, on average, as age increases, the preference for the particular brand of soft drink tends to increase.

It's important to note that correlation does not imply causation. Even though there is a correlation, it doesn't necessarily mean that age causes a preference for a particular soft drink or vice versa. It simply indicates a statistical association between the two variables in the given dataset.

**Q6. A company is interested in examining the relationship between the number of sales calls made per day and the number of sales made per week. The company collected data on both variables from a sample of 30 sales representatives. Calculate the Pearson correlation coefficient between these two variables.**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Example data
data = {'Sales_Calls': [10, 15, 8, 12, 18],
        'Sales_Made': [20, 25, 15, 22, 30]}

# Create a DataFrame
df = pd.DataFrame(data)
print("The Original DataFrame: ")
print(df)

# Extract columns for X and Y
X = df['Sales_Calls']
Y = df['Sales_Made']

# Calculate Pearson correlation coefficient
rho = np.corrcoef(X, Y)[0, 1]
print(f'Pearson Correlation Coefficient: {rho:.2f}')

# Calculate the correlation matrix
correlation_matrix = df.corr()

# Visualize the correlation matrix using seaborn heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1, linewidths=.5)
plt.title('Correlation Matrix')
plt.show()
```

The Pearson correlation coefficient is a measure of the linear relationship between two variables. It ranges from -1 to 1, where:

- 1 indicates a perfect positive linear relationship,
- -1 indicates a perfect negative linear relationship, and
- 0 indicates no linear relationship.

In our case, the Pearson correlation coefficient is 0.99. This value is very close to 1, suggesting a strong positive linear correlation between 'Sales_Calls' and 'Sales_Made' in your dataset.

Interpreting the result:

- As the number of 'Sales_Calls' increases, there is a very high tendency for the 'Sales_Made' to also increase.
- The positive sign (0.99) indicates that as one variable increases, the other variable tends to increase as well.

Keep in mind that correlation does not imply causation. Even though there is a strong correlation, it doesn't necessarily mean that an increase in sales calls causes an increase in sales made or vice versa. Other factors and relationships could be influencing this correlation.
