**Q1. What is the Probability density function?**

A `probability density function (PDF)` describes the likelihood of a `continuous random` variable taking on a specific value. It's often represented by a `continuous curve`, and the area under this curve between two points represents the probability that the random variable falls within that range.

Here's an example in Python using the `normal distribution:`

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Create a range of values
x = np.linspace(-3, 3, 100)

# Define the parameters of the normal distribution (mean and standard deviation)
mean = 0
std_dev = 1.5

# Calculate the PDF values for each x
pdf = norm.pdf(x, loc=mean, scale=std_dev)

# Plot the PDF or probability density function
plt.plot(x, pdf, label='PDF')
plt.xlabel('X')
plt.ylabel('Probability Density')
plt.title('Normal Distribution PDF')
plt.legend()
plt.show()
```

**Q2. What are the types of Probability distribution?**

There are three types of probability distribution function:
- `probability density function (PDF)` for continuous random variable: Some examples of PDFs are normal or Gaussian, log normal, exponential, gamma etc.
- `probability mass function (PMF)` for discrete random variable: Some examples of PMFs are Bernoulli, binomail, uniform, Poisson etc.
- `cumulative distribution function (CDF)`

A `random variable` is a mathematical concept used in probability theory and statistics to describe the possible outcomes of a random experiment or process. It assigns a numerical value to each possible outcome. Random variables can be categorized into two main types: continuous random variables and discrete random variables.
- A `discrete random` variable is one that can take on a countable number of distinct values. These values are often integers and have gaps between them. For example, when `rolling a six-sided die`, the possible outcomes `(1, 2, 3, 4, 5, 6)` are discrete random variables because they are `distinct` and `countable.`
- On the other hand, a `continuous random` variable is one that can take on an infinite number of values within a given range. These values are often represented as real numbers and can be any value within the specified range. An example of a `continuous random` variable is the `height` of individuals in a population, as it can take on any real value within a certain range.

`Examples of different distribution:`

- `Normal Distribution (Gaussian Distribution):` The normal distribution is a bell-shaped curve with a symmetrical and well-defined mean and standard deviation. Example: The distribution of heights in a population.
- `Log-Normal Distribution:` It is a probability distribution of a random variable whose natural logarithm follows a normal distribution. This distribution is commonly used to model data that is positively skewed, such as stock prices, income, and the size of biological populations. Example: Consider the annual returns of a particular stock. If these returns follow a log-normal distribution, it means that the natural logarithm of the annual returns (ln(returns)) follows a normal distribution.
- `Uniform Distribution:`The uniform distribution is characterized by a constant probability. Example: Rolling a fair six-sided die. Each outcome (1 to 6) has an equal probability of 1/6.
- `Bernoulli Distribution:` It is a discrete probability distribution that represents a random experiment with two possible outcomes: success (usually denoted as 1) and failure (usually denoted as 0). It's often used for modeling simple binary events. Example: Suppose we have a biased coin, and we want to model the probability of getting a "heads" (success) when we toss the coin. If the probability of getting heads is p = 0.6, then we can use a Bernoulli distribution to represent the outcomes of each coin toss, where 1 represents "heads" and 0 represents "tails."
- `Binomial Distribution:` The binomial distribution models the number of successful outcomes in a fixed number of independent Bernoulli trials. Example: The probability of getting exactly 3 heads in 5 coin tosses.
- `Poisson Distribution:` The Poisson distribution models the number of events occurring in a fixed interval of time or space. Example: Modeling the number of phone calls received at a call center in an hour.
- `Exponential Distribution:` The exponential distribution models the time between events in a Poisson process. Example: The time between arrivals of customers at a service center.
- `Gamma Distribution:` The gamma distribution is a generalization of the exponential distribution and can be used to model waiting times. Example: Modeling the time until a lightbulb fails.

```python
# Import the required libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, uniform, expon, poisson, lognorm, binom, gamma, bernoulli

# Create a range of values
x = np.linspace(-4, 7, 100)

# Define the parameters for different probability distributions
mean_normal = 0
std_dev_normal = 1.5

low_uniform = 0
high_uniform = 1

lambda_expon = 2

mu_poisson = 3

mu_lognorm = 1
sigma_lognorm = 0.7

n_binom = 10
p_binom = 0.3

alpha_gamma = 2
beta_gamma = 1

p_bernoulli = 0.3

# Calculate the PDF values for each distribution
pdf_normal = norm.pdf(x, loc=mean_normal, scale=std_dev_normal)
pdf_uniform = uniform.pdf(x, loc=low_uniform, scale=high_uniform - low_uniform)
pdf_expon = expon.pdf(x, scale=1/lambda_expon)
pdf_poisson = poisson.pmf(x, mu_poisson)
pdf_lognorm = lognorm.pdf(x, s=sigma_lognorm, loc=0, scale=np.exp(mu_lognorm))
pdf_binom = binom.pmf(x, n_binom, p_binom)
pdf_gamma = gamma.pdf(x, alpha_gamma, loc=0, scale=1/beta_gamma)
pdf_bernoulli = bernoulli.pmf(x, p_bernoulli)

# Plot the PDFs for different probability distributions
plt.figure(figsize=(16, 12))

plt.subplot(3, 3, 1)
plt.plot(x, pdf_normal, label='Normal PDF')
plt.xlabel('X')
plt.ylabel('Probability Density')
plt.title('Normal Distribution PDF')
plt.legend()

plt.subplot(3, 3, 2)
plt.plot(x, pdf_lognorm, label='Log-Normal PDF')
plt.xlabel('X')
plt.ylabel('Probability Density')
plt.title('Log-Normal Distribution PDF')
plt.legend()

plt.subplot(3, 3, 3)
plt.plot(x, pdf_uniform, label='Uniform PDF')
plt.xlabel('X')
plt.ylabel('Probability Density')
plt.title('Uniform Distribution PDF')
plt.legend()

plt.subplot(3, 3, 4)
plt.stem(x, pdf_bernoulli, label='Bernoulli PMF', basefmt=" ")
plt.xlabel('X')
plt.ylabel('Probability Mass')
plt.title('Bernoulli Distribution PMF')
plt.legend()

plt.subplot(3, 3, 5)
plt.stem(x, pdf_binom, label='Binomial PMF', basefmt=" ")
plt.xlabel('X')
plt.ylabel('Probability Mass')
plt.title('Binomial Distribution PMF')
plt.legend()

plt.subplot(3, 3, 6)
plt.stem(x, pdf_poisson, label='Poisson PMF', basefmt=" ")
plt.xlabel('X')
plt.ylabel('Probability Mass')
plt.title('Poisson Distribution PMF')
plt.legend()

plt.subplot(3, 3, 7)
plt.plot(x, pdf_expon, label='Exponential PDF')
plt.xlabel('X')
plt.ylabel('Probability Density')
plt.title('Exponential Distribution PDF')
plt.legend()

plt.subplot(3, 3, 8)
plt.plot(x, pdf_gamma, label='Gamma PDF')
plt.xlabel('X')
plt.ylabel('Probability Density')
plt.title('Gamma Distribution PDF')
plt.legend()

plt.tight_layout()
plt.show()
```

**Q4. What are the properties of Binomial distribution? Give two examples of events where binomial distribution can be applied.**

`Properties of the Binomial Distribution:`

1. **Fixed Number of Trials (n):** The binomial distribution models a fixed number of independent trials, denoted as 'n'.
2. **Two Possible Outcomes:** Each trial has only two possible outcomes, typically labeled as success (S) and failure (F).
3. **Constant Probability of Success (p):** The probability of success (S) remains constant for each trial.
4. **Independent Trials:** Each trial is independent of the others.
5. **Discrete Random Variable:** The binomial random variable represents the number of successful outcomes in 'n' trials.

`Examples of Events where Binomial Distribution can be applied:`

1. **Coin Toss:** When you toss a fair coin multiple times, with each toss having two outcomes (heads or tails) and a constant probability of success (0.5 for a fair coin), you can model the number of heads obtained in 'n' tosses using a binomial distribution.

2. **Quality Control:** In a manufacturing process, if each item produced has a fixed probability of being defective, and you inspect a fixed number of items ('n') for defects, you can model the number of defective items in your sample using a binomial distribution.


- PMF: The probability mass function of a binomial distribution is given by:

$$P(X = k) = \binom{n}{k} p^k (1-p)^{n-k}$$

where $n$ is the number of trials, $k$ is the number of successes, and $p$ is the probability of success in each trial.

- Mean: The mean or expected value of a binomial distribution is given by:

$$\mu = np$$

where $n$ and $p$ are the same as above.

- Median: The median of a binomial distribution is the value of $k$ that satisfies:

$$P(X \leq k) \geq 0.5 \quad \text{and} \quad P(X \geq k) \geq 0.5$$

There is no simple formula for the median, but it can be approximated by:

$$\tilde{\mu} = n p + \frac{1}{2} - \frac{z_{0.5} \sqrt{n p (1-p)}}{n}$$

where $z_{0.5}$ is the 0.5 quantile of the standard normal distribution, which is approximately 0.6745¹.

- Mode: The mode of a binomial distribution is the most likely value of $k$, which is given by:

$$\hat{\mu} = \lfloor (n+1) p \rfloor$$

where $\lfloor x \rfloor$ denotes the largest integer less than or equal to $x$.

- Variance: The variance of a binomial distribution is given by:

$$\sigma^2 = np(1-p)$$

where $n$ and $p$ are the same as above.

- Range: The range of a binomial distribution is the difference between the maximum and minimum possible values of $k$, which are:

$$R = n - 0 = n$$

- Standard deviation: The standard deviation of a binomial distribution is the square root of the variance, which is given by:

$$\sigma = \sqrt{np(1-p)}$$

where $n$ and $p$ are the same as above.

**Q6. Write a Python function to calculate the cumulative distribution function of a Poisson distribution with given mean at a given point.**

The `Poisson` distribution is a `discrete` probability distribution that describes the probability of a given number of events occurring in a fixed interval of time or space, if these events occur with a known constant rate and independently of the time since the last event. For example, the Poisson distribution can be used to model the number of phone calls an office receives during the noon hour, the number of text messages or male grizzly bears per hour, the number of customer arrivals in a cafe per minute, and the number of computer crashes or hospital emergencies per day.

The `probability mass function (PMF)` of the Poisson distribution is given by:

$$P(X=k) = \frac{\lambda^k e^{-\lambda}}{k!}$$

where $X$ is the random variable that counts the number of events, $k$ is a non-negative integer that represents the number of occurrences, $\lambda$ is the mean number of events in the interval, and $e$ is a constant equal to approximately 2.71828.

The PMF gives the probability that exactly $k$ events occur in the interval. For example, if $\lambda = 3$, then $P(X=5) = \frac{3^5 e^{-3}}{5!} \approx 0.1008$, which means that there is about 10.08% chance that exactly 5 events occur in the interval.

The `cumulative distribution function (CDF)` of the Poisson distribution is given by:

$$F(x) = P(X \leq x) = e^{-\lambda} \sum_{j=0}^{\lfloor x \rfloor} \frac{\lambda^j}{j!}$$

where $\lfloor x \rfloor$ is the floor function that rounds down $x$ to the nearest integer.

The CDF gives the probability that at most $x$ events occur in the interval. For example, if $\lambda = 3$, then $F(4) = P(X \leq 4) = e^{-3} \sum_{j=0}^{4} \frac{3^j}{j!} \approx 0.8153$, which means that there is about 81.53% chance that four or less events occur in the interval.

- The `mean` and `variance` of the Poisson distribution are both equal to $\lambda$. This means that the expected value and the measure of dispersion of the number of events are determined by the same parameter. 
- The `standard deviation` is the square root of the variance, so it is equal to $\sqrt{\lambda}$.
- The `mode` of the Poisson distribution is either $\lceil \lambda \rceil - 1$ or $\lfloor \lambda \rfloor$, where $\lceil \lambda \rceil$ is the ceiling function that rounds up $\lambda$ to the nearest integer. The `mode` is the most likely number of events to occur in the interval.
- The `median` of the Poisson distribution is approximately $\lfloor \lambda + \frac{1}{3} - \frac{1}{50\lambda} \rfloor$. The `median` is the middle value of the number of events, such that half of the probabilities are above it and half are below it.
- The `range` of the Poisson distribution is from `0` to `infinity`, since there is no upper limit on how many events can occur in an interval. However, as $k$ increases, the probability of $k$ events occurring decreases exponentially, so most of the probabilities are concentrated near $\lambda$.

`Now first we will see that how the PMF looks like Poission distribution:`

```python
import numpy as np
import matplotlib.pyplot as plt

def poisson_pmf(mean, k):
    """
    Calculate the Probability Mass Function (PMF) of a Poisson distribution with a given mean at a given point.

    :param mean: The mean of the Poisson distribution.
    :param k: The point at which to calculate the PMF.
    :return: The PMF value at the given point.
    """
    return (mean**k * np.exp(-mean)) / np.math.factorial(k)

# Parameters for the Poisson distribution
mean = 8

# Values of k at which to calculate the PMF
k_values = np.arange(0, 11)

# Calculate PMF values at each k and store them in a list
pmf_values = [poisson_pmf(mean, k) for k in k_values]

# Visualize the PMF using matplotlib
plt.bar(k_values, pmf_values, align='center', alpha=0.5)
plt.xlabel('k')
plt.ylabel('PMF')
plt.title('Poisson Distribution PMF')
plt.grid(True)
plt.show()
```

`Now we wiil see the CDF visulization:`

```python
import numpy as np
import matplotlib.pyplot as plt

def poisson_cdf(mean, k):
    """
    Calculate the cumulative distribution function (CDF) of a Poisson distribution with a given mean at a given point.

    :param mean: The mean of the Poisson distribution.
    :param k: The point at which to calculate the CDF.
    :return: The CDF value at the given point.
    """
    cdf = 0
    for i in range(k + 1):
        cdf += np.exp(-mean) * mean ** i / np.math.factorial(i)
    return cdf

# Parameters for the Poisson distribution
mean = 3

# Values of k at which to calculate the CDF
k_values = np.arange(0, 11)

# Calculate CDF values at each k and store them in a list
cdf_values = [poisson_cdf(mean, k) for k in k_values]

# Visualize the CDF using matplotlib
plt.plot(k_values, cdf_values, marker='o', linestyle='-')
plt.xlabel('k')
plt.ylabel('CDF')
plt.title('Poisson Distribution CDF')
plt.grid(True)
plt.show()
```

**Q7. How Binomial distribution different from Poisson distribution?**

**Binomial Distribution:** The Binomial distribution is used when you have a fixed number of trials (n) and each trial can result in one of two outcomes, typically success (S) or failure (F). The probability of success (p) remains constant from trial to trial.

- Example: Tossing a coin (head or tail) multiple times, where the probability of getting a head remains the same in each toss.

**Poisson Distribution:** The Poisson distribution is used when you have a large number of trials (n) and the probability of success (p) is very low, but the average number of successes (λ) is relatively high. This means that, on average, we expect a relatively high number of successful events even though the probability of each individual event is low.

- Example: Counting the number of emails received in an hour, given that the average rate of emails received per hour is low but constant.

**Key Differences:**

- `Number of Trials (n):` In the Binomial distribution, there is a fixed and known number of trials (n). In the Poisson distribution, the number of trials is not fixed and can be very large.
- `Probability of Success (p):` In the Binomial distribution, the probability of success (p) remains constant from trial to trial. In the Poisson distribution, the probability of success (p) is very low but applies to a large number of trials.
- `Outcome Types:` In the Binomial distribution, there are only two possible outcomes (success or failure). In the Poisson distribution, you are counting the number of occurrences of an event within a fixed interval, and there can be multiple outcomes (e.g., the number of emails received).
- `Shape of Distribution:` The Binomial distribution is typically symmetric and bell-shaped when the number of trials is large. The Poisson distribution is often right-skewed and has a longer tail on the right when λ is large.
- `Applicability:` Use the Binomial distribution when you have a fixed number of trials with constant probability. Use the Poisson distribution when the number of trials is large, and the probability is low, but you want to count the number of events in a given interval.

```python
import numpy as np
import matplotlib.pyplot as plt

# Function to calculate Binomial PMF
def binomial_pmf(n, p, k):
    """
    Calculate the Probability Mass Function (PMF) of a Binomial distribution with parameters n and p at a given point k.

    :param n: Number of trials.
    :param p: Probability of success in each trial.
    :param k: The point at which to calculate the PMF.
    :return: The PMF value at the given point.
    """
    binomial_coeff = np.math.comb(n, k)
    return binomial_coeff * p**k * (1 - p)**(n - k)

# Function to calculate Poisson PMF
def poisson_pmf(mean, k):
    """
    Calculate the Probability Mass Function (PMF) of a Poisson distribution with a given mean at a given point.

    :param mean: The mean of the Poisson distribution.
    :param k: The point at which to calculate the PMF.
    :return: The PMF value at the given point.
    """
    return (mean**k * np.exp(-mean)) / np.math.factorial(k)

# Parameters for the Binomial distribution
n = 10  # Number of trials
p = 0.5  # Probability of success

# Parameters for the Poisson distribution
mean = 8

# Values of k for both distributions
k_values = np.arange(0, 11)

# Calculate PMF values for Binomial and Poisson distributions at each k
pmf_values_binomial = [binomial_pmf(n, p, k) for k in k_values]
pmf_values_poisson = [poisson_pmf(mean, k) for k in k_values]

# Create a figure with two subplots side by side
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Plot Binomial PMF
ax1.bar(k_values, pmf_values_binomial, align='center', alpha=0.5, label='Binomial', color='blue')
ax1.set_xlabel('k')
ax1.set_ylabel('PMF')
ax1.set_title('Binomial Distribution PMF')
ax1.grid(True)
ax1.legend()

# Plot Poisson PMF
ax2.bar(k_values, pmf_values_poisson, align='center', alpha=0.5, label='Poisson', color='red')
ax2.set_xlabel('k')
ax2.set_ylabel('PMF')
ax2.set_title('Poisson Distribution PMF')
ax2.grid(True)
ax2.legend()

plt.tight_layout()
plt.show()
```

**Q8. Generate a random sample of size 1000 from a Poisson distribution with mean 5 and calculate the sample mean and variance.**

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a random sample of size 1000 from a Poisson distribution with mean 5
sample_size = 1000
mean = 5
poisson_sample = np.random.poisson(mean, size=sample_size)

# Calculate the sample mean and variance
sample_mean = np.mean(poisson_sample)
sample_variance = np.var(poisson_sample)

# Visualize the PMF of the Poisson distribution
k_values = np.arange(0, 15)  # Define a range of k values for the PMF
pmf_values = [np.sum(poisson_sample == k) / sample_size for k in k_values]

plt.bar(k_values, pmf_values, align='center', alpha=0.5, label='Poisson', color='green')
plt.xlabel('k')
plt.ylabel('PMF')
plt.title(f'Poisson Distribution PMF (Sample Size = {sample_size})')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()

# Output the sample mean and variance
print(f"Sample Mean: {sample_mean}")
print(f"Sample Variance: {sample_variance}")
```

**Q9. How mean and variance are related in Binomial distribution and Poisson distribution?**

**Binomial Distribution:**

`Mean (μ) = n * p`

- Where "n" is the number of trials.
- "p" is the probability of success in each trial.
- Variance (σ^2) = n * p * (1 - p)

In a `Binomial distribution`, the variance is directly proportional to the product of the number of trials (n) and the probability of success (p) and inversely proportional to the probability of failure (1 - p). This means that as either "n" or "p" increases, the variance also increases.

**Poisson Distribution:**

`Mean (μ) = λ`

- Where "λ" is the mean of the Poisson distribution.
- Variance (σ^2) = λ

In a `Poisson distribution`, the variance is equal to the mean (λ). This relationship implies that the variance is constant and does not depend on any other parameters. It's a unique characteristic of the Poisson distribution that the mean and variance are equal.

- In summary, for a `Binomial distribution`, the variance depends on both the number of trials and the probability of success, while for a Poisson distribution, the variance is solely determined by the mean, and it is equal to the mean.

**Q10. In normal distribution with respect to mean position, where does the least frequent data appear?**

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters of the normal distribution
mean = 0  # Mean position
std_dev = 1  # Standard deviation

# Generate x values for the probability density function (PDF)
x = np.linspace(mean - 3 * std_dev, mean + 3 * std_dev, 1000)

# Calculate the PDF using the probability density function formula
pdf = 1 / (std_dev * np.sqrt(2 * np.pi)) * np.exp(-(x - mean) ** 2 / (2 * std_dev ** 2))

# Visualize the PDF
plt.plot(x, pdf, label='PDF')
plt.title('Probability Density Function of a Normal Distribution')
plt.xlabel('X')
plt.ylabel('Probability Density')
plt.grid(True)
plt.legend()
plt.show()
```

In a `normal` distribution, the `least frequent data` appears in the tails, which are the `extreme positive` and `negative` values, far from the `mean `position.
