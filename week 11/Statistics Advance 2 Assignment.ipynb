**Q1: What are the Probability Mass Function (PMF) and Probability Density Function (PDF)? Explain with an example.**

**Probability Mass Function (PMF):** The PMF is a function used in the context of discrete random variables. It assigns a probability to each possible outcome of the random variable. In simpler terms, it tells us the likelihood of each specific value occurring.

**Example:** Suppose we are rolling a fair six-sided die. The PMF for this situation would be:

- P(X = 1) = 1/6
- P(X = 2) = 1/6
- P(X = 3) = 1/6
- P(X = 4) = 1/6
- P(X = 5) = 1/6
- P(X = 6) = 1/6

Here, the PMF gives the probability of each outcome (each number on the die) occurring.

**Probability Density Function (PDF):** The PDF is used for continuous random variables. It describes the probability distribution of a continuous random variable, and the area under the PDF curve within a given range represents the probability of the variable falling within that range.

**Example:** Let's consider a uniform distribution between 0 and 1. The PDF would be a straight line from 0 to 1, with constant height, because all values in that range are equally likely. The area under this line between two points, say 0.2 and 0.5, represents the probability of the variable falling in that range.

Now, let's move on to the code to visualize PMF and PDF side by side. We'll use Python and the matplotlib library to create the plot:

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import uniform

# Discrete example data
x_pmf = [1, 2, 3, 4, 5, 6]
pmf_values = [1/6, 1/6, 1/6, 1/6, 1/6, 1/6]

# Continuous example data
x_pdf = np.linspace(-1, 2, 100)
# Define the interval [a, b]
a, b = 0, 1
# Create an instance of the uniform distribution class
# The second argument is the scale parameter, which is b-a
u = uniform(a, b-a)
# Calculate the PDF values
pdf_values = u.pdf(x_pdf)

# Create a subplot with two plots side by side
plt.figure(figsize=(12, 4))

# Plot the PMF
plt.subplot(1, 2, 1)
plt.bar(x_pmf, pmf_values, alpha=0.5, align='center')
plt.xlabel("Outcome")
plt.ylabel("Probability")
plt.title("Probability Mass Function (PMF)")

# Plot the PDF
plt.subplot(1, 2, 2)
plt.plot(x_pdf, pdf_values, label='Uniform PDF')
plt.xlabel("Value")
plt.ylabel("Probability Density")
plt.title("Probability Density Function (PDF)")

# Adjust layout for better readability
plt.tight_layout()
plt.show()
```

**Q2: What is Cumulative Density Function (CDF)? Explain with an example. Why CDF is used?**

Cumulative Density Function (CDF) is a function used to describe the probability that a continuous random variable will take a value less than or equal to a given value. In the case of discrete random variables, it sums up the probabilities of all values less than or equal to a given value.

Example:
Let's consider a continuous random variable X, uniformly distributed between 0 and 1. 

`The CDF for this situation is defined: CDF(X <= x) = x for 0 <= x <= 1.`

- This means that the probability of X being less than or equal to a specific value x is equal to x for all x between 0 and 1. It's a linear function that starts at 0 when x is 0 and reaches 1 when x is 1, which is why it's a straight line with a slope of 1.
- This makes intuitive sense because for a uniform distribution, the probability is equally distributed across the entire range from 0 to 1. So, as 'x' increases from 0 to 1, the cumulative probability also increases linearly.
- CDF is used to answer questions like "What is the probability that X is less than or equal to a certain value?" and is particularly useful in probability and statistics to analyze and work with continuous random variables.

`If we want to visualize the CDF for the Uniform PDF:`

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import uniform

# Continuous example data
x_pdf = np.linspace(-1, 2, 100)
# Define the interval [a, b]
a, b = 0, 1
# Create an instance of the uniform distribution class
# The second argument is the scale parameter, which is b-a
u = uniform(a, b-a)
# Calculate the PDF values
pdf_values = u.pdf(x_pdf)

# Create a subplot with two plots side by side
plt.figure(figsize=(12, 4))

# Plot the PDF
plt.subplot(1, 2, 1)
plt.plot(x_pdf, pdf_values, label='Uniform PDF')
plt.xlabel("Value")
plt.ylabel("Probability Density")
plt.title("Probability Density Function (PDF)")

# Calculate the CDF values for the uniform distribution
cdf_values = u.cdf(x_pdf)

# Add a subplot for the CDF
plt.subplot(1, 2, 2)
plt.plot(x_pdf, cdf_values, label='Uniform CDF', color='red')
plt.xlabel("Value")
plt.ylabel("Cumulative Probability")
plt.title("Cumulative Density Function (CDF)")

# Adjust layout for better readability
plt.tight_layout()
plt.show()
```

**Q3: What are some examples of situations where the normal distribution might be used as a model? Explain how the parameters of the normal distribution relate to the shape of the distribution.**

**Here are some examples of situations where the normal distribution might be used as a model:**

1. `Heights of a Population:` The normal distribution is often used to model the heights of a population, where the mean and standard deviation correspond to the average height and the variation in heights, respectively.
2. `Test Scores:` Test scores, like SAT or IQ scores, are often modeled using a normal distribution. The mean represents the average score, and the standard deviation accounts for the spread of scores.
3. `Errors in Measurements:` Measurement errors, such as errors in readings from instruments, can be modeled with a normal distribution, where the mean is the true value, and the standard deviation represents the measurement error.
4. `Natural Phenomena:` Many natural phenomena, like the distribution of particle speeds in a gas or the heights of ocean waves, can be approximated by the normal distribution.

**Parameters of the normal distribution:**

- `Mean ():` The mean determines the center of the distribution, and it corresponds to the highest point on the normal curve.
- `Standard Deviation ():` The standard deviation controls the spread or dispersion of the distribution. A larger standard deviation results in a wider and flatter curve.

`Example of creating and plotting a normal distribution:`

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

# Generate data for the normal distribution with mean 0 and standard deviation 1
mu = 0  # Mean
sigma = 1  # Standard Deviation
x = np.linspace(-5, 5, 100)
pdf_values = norm.pdf(x, mu, sigma)  # PDF values

# Generate data for another normal distribution with a higher standard deviation
sigma_high = 2  # Higher Standard Deviation
pdf_values_high = norm.pdf(x, mu, sigma_high)  # PDF values for the higher standard deviation

# Create subplots
plt.figure(figsize=(12, 6))

# Subplot for the normal distribution with standard deviation 1
plt.subplot(1, 2, 1)
plt.plot(x, pdf_values, label='Std Deviation 1', color='green')
plt.xlabel("Value")
plt.ylabel("Probability Density")
plt.title("Normal Distribution (=1)")
plt.legend()
plt.grid(True)

# Subplot for the normal distribution with a higher standard deviation
plt.subplot(1, 2, 2)
plt.plot(x, pdf_values_high, label='Std Deviation 2', color='blue')
plt.xlabel("Value")
plt.ylabel("Probability Density")
plt.title("Normal Distribution (=2)")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

**Q5: What is Bernaulli Distribution? Give an Example. What is the difference between Bernoulli
Distribution and Binomial Distribution?**

**Bernoulli Distribution:**
The Bernoulli distribution is a discrete probability distribution that models a single trial of a binary outcome, which can be either a success (usually denoted as 1) or a failure (usually denoted as 0). It is characterized by a single parameter, often denoted as p, which represents the probability of success in a single trial.

**Example:**
A classic example of the Bernoulli distribution is modeling the outcome of a coin toss. If you define "heads" as a success (1) and "tails" as a failure (0), the probability of getting a head (success) in a single coin toss would be the parameter p.

**Difference between Bernoulli Distribution and Binomial Distribution:**

1. **Number of Trials:**
   - Bernoulli Distribution: Models a single trial (one attempt) with a binary outcome.
   - Binomial Distribution: Models multiple independent and identical Bernoulli trials (repeated attempts) and counts the total number of successes.

2. **Parameter:**
   - Bernoulli Distribution: Has a single parameter p, representing the probability of success in a single trial.
   - Binomial Distribution: Has two parameters, n (number of trials) and p (probability of success in a single trial), where n > 1.

3. **Outcome:**
   - Bernoulli Distribution: Produces either a 1 (success) or a 0 (failure) in a single trial.
   - Binomial Distribution: Produces the count of successes (0, 1, 2, ..., n) in a fixed number of trials.

4. **Purpose:**
   - Bernoulli Distribution: Used to model individual binary events or trials, such as a single coin toss.
   - Binomial Distribution: Used to model the number of successes in a fixed number of independent Bernoulli trials, like the number of heads in 10 coin tosses.

5. **Probability Function:**
   - Bernoulli Distribution: Probability mass function (PMF) for a single trial is P(X = 1) = p and P(X = 0) = 1 - p.
   - Binomial Distribution: PMF for the number of successes (k) in n trials is given by the binomial coefficient and the individual trial's probability: P(X = k) = C(n, k) * p^k * (1-p)^(n-k), where C(n, k) is the binomial coefficient.

In probability and combinatorics, `C(n, k)`, often denoted as `n choose k` or "combinations," represents the binomial coefficient. It is a way to calculate the number of ways to choose k items from a set of n items without regard to the order in which they are chosen. The formula for the binomial coefficient is:

$$C(n, k) = \binom{n}{k} = \frac{n!}{k!(n-k)!}$$

Where:
- n is the total number of items in the set.
- k is the number of items to choose from the set.
- "!" denotes the factorial of a number, which is the product of all positive integers from 1 to that number.

In the context of the `binomial distribution, C(n, k)` is used to calculate the number of ways to obtain exactly `k successes` in `n independent Bernoulli trials`, each with a probability of success denoted as `p.` It's a fundamental concept in probability and combinatorics and plays a significant role in calculating probabilities for the binomial distribution.

In summary, the Bernoulli Distribution models a single trial, while the Binomial Distribution extends this to model the number of successes in a fixed number of independent trials.

**Q6. Consider a dataset with a mean of 50 and a standard deviation of 10. If we assume that the dataset is normally distributed, what is the probability that a randomly selected observation will be greater than 60? Use the appropriate formula and show your calculations.**

**Answer:** To calculate the probability that a randomly selected observation from a normally distributed dataset with a mean of 50 and a standard deviation of 10 will be greater than 60, we can use the z-score formula and the standard normal distribution (z-distribution).

The formula to calculate the z-score is:

$$Z = \frac{X - \mu}{\sigma}$$

Where:

- $Z$ is the z-score.
- $X$ is the value we want to find the probability for (in this case, 60).
- $\mu$ is the mean of the dataset (50).
- $\sigma$ is the standard deviation of the dataset (10).

First, calculate the z-score for (X = 60):

$$Z = \frac{60 - 50}{10} = 1 $$

Now, we can find the probability using a z-table or a calculator. The probability that a randomly selected observation will be greater than 60 is the area under the standard normal distribution curve to the right of (Z = 1).

Using a z-table or calculator, we can find (P(Z > 1)), which is the probability that (X > 60). This value represents the probability that a randomly selected observation will be greater than 60.

Using the Z table, the probability value corresponding to row value 1.0 and column value 0.00 is 0.84134. So, the probability that (X > 60) is:

$$P(X > 60) = 1 - P(X < 60) = 1 - 0.84134 = 0.15866.$$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters
mean = 50
std_dev = 10
X = 60

# Calculate the z-score for X
z = (X - mean) / std_dev

# Create a range of x values
x = np.linspace(mean - 3 * std_dev, mean + 3 * std_dev, 1000)

# Calculate the probability density function (PDF) for the normal distribution
pdf = norm.pdf(x, loc=mean, scale=std_dev)

# Calculate the probability P(X > 60)
probability = 1 - norm.cdf(X, loc=mean, scale=std_dev)

# Create a shaded area to represent the probability P(X > 60)
x_shaded = np.linspace(X, mean + 3 * std_dev, 1000)
pdf_shaded = norm.pdf(x_shaded, loc=mean, scale=std_dev)

# Plot the normal distribution and shaded area
plt.plot(x, pdf, label="Normal Distribution")
plt.fill_between(x_shaded, pdf_shaded, color='red', alpha=0.5, label=f"P(X > {X}) = {probability:.5f}")

# Annotate the z-score
plt.annotate(f'Critical Z Score = {z:.2f}', xy=(X, 0.02), xytext=(X , 0.03),
             arrowprops=dict(arrowstyle='->'))

# Add labels and legend
plt.xlabel("X")
plt.ylabel("Probability Density")
plt.title("Probability Visualization")
plt.legend()

# Show the plot
plt.show()
```

**Q7: Explain uniform Distribution with an example.** 

The `uniform distribution` is a continuous probability distribution in which all values within a given range are equally likely to occur. It is often denoted as `U(a, b)` where `a` is the lower bound and `b` is the upper bound of the distribution.

`Probability Density Function (PDF):` The probability density function for the uniform distribution is defined as:

$$f(x) = \frac{1}{b - a} \text{ for } a \leq x \leq b$$

$$f(x) = 0 \text{ for } x < a \text{ or } x > b$$

Where:

- $a$ is the lower bound.
- $b$ is the upper bound.
- The PDF is a constant value within the range `((a, b)).`

`Cumulative Distribution Function (CDF):` The cumulative distribution function for the uniform distribution is defined as:

$$F(x) = \frac{x - a}{b - a} \text{ for } a \leq x \leq b$$

$$F(x) = 0 \text{ for } x < a$$

$$F(x) = 1 \text{ for } x \geq b$$

Where:

- $a$ is the lower bound.
- $b$ is the upper bound.

`Mean (Expected Value):` The mean of a uniform distribution is the average of the lower and upper bounds:

$$\mu = \frac{a + b}{2}$$

`Mode:` The mode of a uniform distribution is any value within the range `((a, b)).`

`Median:` The median of a uniform distribution is also any value within the range `((a, b)).`

`Variance:` The variance of a uniform distribution is calculated as:

$$\sigma^2 = \frac{(b - a)^2}{12} $$

`Range:` The range is the difference between the upper and lower bounds:

$$\text{Range} = b - a $$

`Standard Deviation:` The standard deviation is the square root of the variance:

$$\sigma = \sqrt{\frac{(b - a)^2}{12}}$$

Here are some real-life examples where the uniform distribution is applied:

- `Random Number Generation:` When you roll a fair six-sided die, the outcomes (1, 2, 3, 4, 5, 6) follow a uniform distribution because each outcome has an equal probability of 1/6.

- `Lottery Draw:` In some lotteries, a set of numbers is drawn randomly from a uniform distribution, where each possible number has an equal chance of being drawn.

- `Quality Control:` When inspecting items on a production line, if defects are equally likely to occur at any point in the process, the distribution of defects may follow a uniform distribution.

- `Weather Forecasting:` In some climate models, the distribution of daily temperatures in a specific region may be approximated using a uniform distribution within a certain range.

- `Randomized Experiments:` In clinical trials or scientific experiments, randomization of subjects or conditions can be done using a uniform distribution to ensure fairness.

- `Traffic Signal Timing:` The timing of traffic signals at intersections can be optimized using a uniform distribution to allocate green and red light times fairly to different directions.

`We can also visualize the uniform distribution:`

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import uniform

# Continuous example data
x_pdf = np.linspace(-1, 2, 100)
# Define the interval [a, b]
a, b = 0, 1
# Create an instance of the uniform distribution class
# The second argument is the scale parameter, which is b-a
u = uniform(a, b-a)
# Calculate the PDF values
pdf_values = u.pdf(x_pdf)

# Plot the PDF
plt.plot(x_pdf, pdf_values, label='Uniform PDF')
plt.xlabel("Value")
plt.ylabel("Probability Density")
plt.title("Probability Density Function (PDF)")

# Adjust layout for better readability
plt.tight_layout()
plt.show()
```

**Q8: What is the z score? State the importance of the z score.**

`Z-score` is a measure of how many standard deviations a data point is from the `mean` of a dataset. It's important because it allows us to standardize and compare data points from different distributions.

`Properties of Z score:`

- The Z-score tells us how many standard deviations the data point is from the mean.
- If z_score is negative, it means the data point is below the mean, and if it's positive, it's above the mean.
- A Z-score of 0 means the data point is at the mean.

Calculate the Z-score for a data point `x` in a dataset with mean `mu` and standard deviation `sigma`:

```python
def calculate_z_score(x, mu, sigma):
    z = (x - mu) / sigma
    return z

# Example usage:
x = 68  # Your data point
mu = 70  # Mean of the dataset
sigma = 3  # Standard deviation of the dataset

z_score = calculate_z_score(x, mu, sigma)
print(f"The Z-score for {x} is {z_score}")
```

`Importance of Z-score:`

1. Standardization: Z-scores make it possible to standardize data, allowing comparisons across different datasets.
2. Outlier Detection: Z-scores help identify outliers, which are data points far from the mean.
3. Normal Distribution: Z-scores are used to assess normality and check if data follows a normal distribution.
4. Hypothesis Testing: Z-scores are essential in hypothesis testing and determining statistical significance.

**Q9: What is Central Limit Theorem? State the significance of the Central Limit Theorem.**

`Central Limit Theorem (CLT)` is a fundamental concept in statistics. It states that the distribution of the sample means of a sufficiently large number of independent, identically distributed random variables will be approximately normally distributed, regardless of the original population's distribution.

$\text{Let } X_1, X_2, \dots, X_n \text{ be independent and identically distributed random variables with mean } \mu \text{ and standard deviation } \sigma. \text{ Then, as } n \text{ approaches infinity, the sampling distribution of the sample mean } \overline{X} \text{ is approximately normal with mean } \mu \text{ and standard deviation } \frac{\sigma}{\sqrt{n}}.\text{ That is, }\overline{X} \sim N\left(\mu, \frac{\sigma}{\sqrt{n}}\right)$


`Significance of the Central Limit Theorem:`
1. Normality of Sample Means: The CLT ensures that, for a large sample size, the distribution of sample means will be approximately normal, even if the population is not normally distributed. This is crucial for making inferences about the population.
2. Estimation: It allows us to use the properties of the normal distribution to estimate population parameters based on sample statistics.
3. Hypothesis Testing: The CLT is the foundation for many hypothesis tests and confidence intervals in statistics.
4. Generalizability: It enables us to make generalizations from sample data to the entire population, as long as the sample size is sufficiently large.
5. Practical Application: The CLT is widely used in various fields, such as quality control, economics, and social sciences, for making informed decisions based on sample data.

It's important to note that the CLT applies to the distribution of sample means, not individual observations, and the sample size must be sufficiently large for the approximation to hold.

**Q10: State the assumptions of the Central Limit Theorem.**

`Here are the assumptions of the Central Limit Theorem (CLT):`

- `Independence:` The random variables in the sample should be independent of each other. This means that the value of one random variable should not provide any information about the value of another random variable in the sample.

- `Identical Distribution:` All random variables in the sample should be drawn from the same probability distribution. In other words, they should have the same mean () and standard deviation ().

- `Sample Size:` The sample size (n) should be sufficiently large. While there isn't a fixed rule for what constitutes "sufficiently large," a common guideline is that `n should be greater than or equal to 30.` However, the exact threshold may vary depending on the original distribution of the population.

These assumptions are essential for the Central Limit Theorem to hold, and they enable the sampling distribution of the sample means to be approximately normal, regardless of the shape of the original population's distribution.
