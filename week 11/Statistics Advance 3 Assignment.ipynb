**Q1: What is Estimation Statistics? Explain point estimate and interval estimate.**

`Estimation Statistics` involves using sample data to make inferences about population parameters. There are two main types of estimation:

- `Point Estimate:` A point estimate is a single value that is used to estimate an unknown population parameter. It's a best guess or an approximation based on a sample of data. For example, if we want to estimate the average height of all students in a school, we might take a random sample of students and calculate the mean height of that sample. This mean height is our point estimate for the population's average height.

- `Interval Estimate:` An interval estimate, on the other hand, provides a range or interval of values within which the population parameter is likely to fall. This interval is associated with a certain `level of confidence.` It takes into account the uncertainty inherent in estimating a population parameter from a sample. A common form of interval estimate is a `confidence interval`, which provides a range of values and a confidence level. For example, we might estimate the average income of a population as 40,000 with a confidence level 95 is `$38,000` to `$42,000.` This means we are 95% confident that the true average income falls within this range.

**Q2. Write a Python function to estimate the population mean using a sample mean and standard deviation.**

```python
def estimate_population_mean(sample_mean, sample_std_dev, sample_size):
    """
    Estimate the population mean using a sample mean and standard deviation.
    
    :param sample_mean: The mean of the sample data.
    :param sample_std_dev: The standard deviation of the sample data.
    :param sample_size: The size of the sample.
    :return: Estimated population mean.
    """
    import math
    
    # Calculate the standard error
    standard_error = sample_std_dev / math.sqrt(sample_size)
    
    # Calculate the margin of error for a certain confidence level (e.g., 95%)
    confidence_level = 0.95  # You can adjust this for different confidence levels
    z_score = 1.96  # Z-score for 95% confidence interval
    margin_of_error = z_score * standard_error
    
    # Calculate the lower and upper bounds of the confidence interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error
    
    return lower_bound, upper_bound

# Example usage:
sample_mean = 500
sample_std_dev = 50
sample_size = 100
lower, upper = estimate_population_mean(sample_mean, sample_std_dev, sample_size)
print(f"Estimated population mean is between {lower} and {upper}")
```

**Q3: What is Hypothesis testing? Why is it used? State the importance of Hypothesis testing.**

`Hypothesis testing` is a statistical method used to make inferences or draw conclusions about a population based on a sample of data. It is an essential tool in scientific research and decision-making in various fields, including science, business, and healthcare.

The importance of hypothesis testing lies in its ability to:

- `Test Assumptions:` Hypothesis testing helps in evaluating and validating assumptions or claims about a population. For example, a pharmaceutical company might want to test whether a new drug is more effective than an existing one.
- `Quantify Uncertainty:` It provides a structured way to quantify the uncertainty associated with sample data and make decisions based on this uncertainty.
- `Support Decision-Making:` It helps in making informed decisions by providing evidence to accept or reject a hypothesis. For instance, in quality control, it can be used to decide whether a manufacturing process is producing items within the specified tolerances.
- `Generalization:` It allows generalizing the results from a sample to the entire population. This is crucial when it's impractical to study an entire population.
- `Scientific Rigor:` In scientific research, it ensures that conclusions are drawn based on statistical evidence rather than mere intuition or anecdotal evidence.
- `Quality Control:` In industries, it plays a vital role in ensuring the quality and consistency of products.

`Example:` Consider a company that claims their new advertising campaign has increased sales. The null hypothesis (H0) could be `the new campaign has no effect on sales`, and the alternative hypothesis (H1) could be `the new campaign has increased sales.` Hypothesis testing allows us to determine whether there is enough evidence to reject the null hypothesis in favor of the alternative, thus providing a basis for making decisions about the effectiveness of the advertising campaign.

**Q5. Write a Python script to conduct a hypothesis test on the difference between two population means, given a sample from each population.**

**Independent Two-Sample T-Test:**

The `independent two-sample t-test` is a statistical test used to determine if there is a significant difference between the means of two independent population groups. It is often applied when we have two groups, and we want to test whether the means of a particular variable are different between these groups. 
- The null hypothesis states that there is no significant difference between the two groups,
- while the alternative hypothesis suggests that there is a significant difference.

**T-Statistic Formula:** The formula for the t-statistic is given as:

$$t = \frac{{\bar{x}_1 - \bar{x}_2}}{{\sqrt{\frac{s_1^2}{n_1} + \frac{s_2^2}{n_2}}}}$$

Where:

- $(\bar{x}_1)$ and $(\bar{x}_2)$ are the sample means of the two poulation groups.
- $(s_1)$ and $(s_2)$ are the sample standard deviations of the two population groups.
- $(n_1)$ and $(n_2)$ are the sample sizes of the two population groups.

`Now we can see the below example for this test:`

```python
import scipy.stats as stats

# Define our data for two populations
# Sample data for population 1 (e.g., male students' weights)
population1 = [175, 180, 185, 170, 172] 
# Sample data for population 2 (e.g., female students' weights)
population2 = [160, 165, 162, 168, 175] 
# Calculate the t-statistic and p-value
t_statistic, p_value = stats.ttest_ind(population1, population2)

# Define the significance level (alpha)
alpha = 0.05

# Check if the p-value is less than alpha to determine significance
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference between the two populations.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference between the two populations.")
```

**Q7: Write down the steps involved in hypothesis testing.**

**Steps Involved in Hypothesis Testing:**

Hypothesis testing is a statistical method used to make inferences about a population based on a sample of data. It typically involves the following steps:

1. **Formulate Hypotheses:**
   - State the null hypothesis $(H_0)$ and the alternative hypothesis $(H_a)$.
   - $H_0$ represents the default assumption, often stating that there is no effect or no difference, while $H_a$ suggests there is an effect or a difference.
   - For example, if we are testing a new drug's effectiveness:
     - $H_0$: The new drug has no effect on patient recovery.
     - $H_a$: The new drug has a positive effect on patient recovery.

2. **Collect Data:**
   - Gather a sample of data that is relevant to your hypothesis.

3. **Choose a Significance Level $(\alpha)$:**
   - Set the significance level, denoted by $(\alpha)$, which represents the probability of making a Type I error (rejecting $H_0$ when it's true).
   - Common values for $\alpha$ are 0.05 and 0.01.

4. **Perform a Statistical Test:**
   - Select an appropriate statistical test based on the type of data and the hypothesis being tested.
   - Calculate the test statistic like Z, t and its associated p-value.

5. **Make a Decision:**
   - Compare the p-value to the significance level $(\alpha)$.
   - If the p-value is less than or equal to $\alpha$, reject the null hypothesis.
   - If the p-value is greater than $\alpha$, fail to reject the null hypothesis.

6. **Draw a Conclusion:**
   - Based on our decision in step 5, draw a conclusion about the null hypothesis.
   - If we reject $H_0$, we can conclude that there is evidence in favor of $H_a$.
   - If we fail to reject $H_0$, we can conclude that there is not enough evidence to support $H_a$.

7. **Report Results:**
   - Communicate our findings, including the test statistic, p-value, and the conclusion in a clear and concise manner.

8. **Consider Practical Significance:**
   - Evaluate the practical significance of the results, not just statistical significance. Consider whether the observed effect is meaningful in the real world.

9. **Repeat and Verify:**
   - In some cases, it's advisable to replicate the study or conduct further research to ensure the results are consistent.

**Q8. Define p-value and explain its significance in hypothesis testing.**

In `hypothesis testing`, the `p-value` is a crucial statistical measure that helps us determine the strength of evidence against the null hypothesis ($H_0$). The p-value quantifies the probability of obtaining a test statistic as extreme as, or more extreme than, the one observed in our sample, assuming that the null hypothesis is true. In other words, it tells us how likely we would see the observed data if the null hypothesis were correct.

`The significance of the p-value can be summarized as follows:`
- A small p-value (typically less than the chosen significance level $\alpha$, e.g., 0.05) indicates strong evidence against the null hypothesis. This suggests that the observed data is unlikely to have occurred by random chance, and we may reject the null hypothesis in favor of the alternative hypothesis ($H_a$).
- A large p-value, on the other hand, suggests weak evidence against the null hypothesis. It implies that the observed data is consistent with what we would expect if $H_0$ were true. In this case, we fail to reject the null hypothesis.

`To illustrate, let's consider an example:`

Suppose we are testing the effectiveness of a new drug, and our `null hypothesis is that the drug has no effect on patient recovery ($H_0$).` After conducting the test, we calculate a `p-value of 0.03`, which is less than our chosen significance level of `0.05.` This small p-value suggests that there is strong evidence that the new drug has a positive effect on patient recovery, and we `reject the null hypothesis` in favor of the `alternative hypothesis.`

**Q10. Write a Python program to calculate the two-sample t-test for independent samples, given two random samples of equal size and a null hypothesis that the population means are equal.**

```python
import numpy as np
from scipy import stats

# Define two random samples
sample1 = np.array([28, 32, 35, 38, 26, 30, 34, 29, 31, 27])
sample2 = np.array([36, 39, 41, 37, 33, 40, 35, 42, 38, 36])

# Define the null hypothesis (H0): Population means are equal
# You can change this null hypothesis as needed
null_hypothesis = "Population means are equal"

# Perform the two-sample t-test
t_statistic, p_value = stats.ttest_ind(sample1, sample2)

# Define the significance level (alpha)
alpha = 0.05

# Check if the p-value is less than alpha to determine the result
if p_value < alpha:
    result = "Reject the null hypothesis"
else:
    result = "Fail to reject the null hypothesis"

# Print the results
print("Null Hypothesis:", null_hypothesis)
print("Sample 1:", sample1)
print("Sample 2:", sample2)
print("T-Statistic:", t_statistic)
print("P-Value:", p_value)
print("Result:", result)
```

**Q13. A coffee shop owner wants to estimate the average daily revenue for their shop. They take a random sample of 50 days and find the sample mean revenue to be `$500` with a standard deviation of `$50.` Estimate the population mean revenue with a 95% confidence interval.**

To estimate the population mean revenue with a 95% confidence interval, we can use the **Z-test**. Here are the steps:

1. **Symbol Definitions**:
   - Population mean revenue: $\mu$
   - Sample mean revenue: $\bar{x} = 500$
   - Population standard deviation: $\sigma = 50$
   - Sample size: $n = 50$
   - Confidence level: (1 - $\alpha$) = 0.95
   - Therefore: $\alpha$ = 1 - 0.95 = 0.05
   
2. **Critical Regions and Alpha Level:**

   - Critical regions are ranges of the test statistic’s distribution where values lead to statistically significant results.
   - The alpha level (significance level) determines the size of these critical regions.
   - For a two-tailed test, we split the alpha level into both tails (left and right).
   - For a one-tailed test, the entire alpha level is in one tail.

3. **Determining One-Tailed vs. Two-Tailed:**

   - If it specifies a direction (e.g., “greater than,” “less than”), use a one-tailed test.
   - If it doesn’t specify a direction, use a two-tailed test.  
   - No specific direction is mentioned, so it’s a two-tailed test.
   - We use $Z_{\alpha/2}$ to find critical values for both tails.

4. **Calculate the Margin of Error:**

   - The margin of error $E$ is given by:
   
$$E = Z_{\alpha/2} \cdot \frac{\sigma}{\sqrt{n}}$$

   - where $Z_{\alpha/2}$ is the critical Z-score corresponding to the desired confidence level. For a 95% confidence interval, $Z_{\alpha/2} = 1.96$ (approximately).

5. **Confidence Interval**:
   - The confidence interval for the population mean revenue is:
   
     $$(\bar{x} - E, \bar{x} + E)$$

6. **Substitute Values**:
   - we can calculate:
   
     $$E = 1.96 \cdot \frac{50}{\sqrt{50}} \approx 13.94$$
     
   - So the confidence interval is approximately:
   
     $$(500 - 13.94, 500 + 13.94) = (486.06, 513.94)$$

7. **Conclusion**:
   - We are 95% confident that the true population mean revenue falls within the interval (486.06, 513.94).

Remember that this interval represents our estimate based on the sample data, and there is a 95% chance that it contains the true population mean revenue. If we were to repeat this process with different samples, we would expect approximately 95% of those intervals to contain the true population mean revenue.

`We acn see the visualization in below:`

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Define the sample statistics
sample_mean = 500
population_std = 50
sample_size = 50

# Define the confidence level
confidence_level = 0.95

# Step 4: Calculate the margin of error (E)
# Calculate the critical Z-score for a 95% confidence interval
alpha = 1 - confidence_level
z_alpha_over_2 = norm.ppf(1 - alpha / 2)

# Calculate the margin of error
margin_of_error = z_alpha_over_2 * (population_std / np.sqrt(sample_size))

# Step 5: Calculate the confidence interval
lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

# Step 7: Print the results
print(f"We are {confidence_level * 100}% confident that the true population mean revenue falls within the interval "
      f"({lower_bound:.2f}, {upper_bound:.2f}).")

# Step 8: Visualize the data
x = np.linspace(sample_mean - 3 * population_std, sample_mean + 3 * population_std, 1000)
y = norm.pdf(x, sample_mean, population_std)

plt.figure(figsize=(10, 5))
plt.plot(x, y, label='Normal Distribution', color='b')
plt.fill_between(x, 0, y, where=(x >= lower_bound) & (x <= upper_bound), color='yellow', alpha=0.5,
                 label='95% Confidence Interval')
plt.axvline(sample_mean, color='r', linestyle='--', label='Sample Mean')
# Add arrow and text box to display the range values on the x-axis
plt.annotate(f'lower_bound = {lower_bound:.2f}', xy=(lower_bound, 0), xytext=(350, 0.004),
             arrowprops=dict(arrowstyle='->', color='green'), fontsize=12)
plt.annotate(f'upper_bound = {upper_bound:.2f}', xy=(upper_bound, 0), xytext=(570, 0.004),
             arrowprops=dict(arrowstyle='->', color='green'), fontsize=12)
plt.legend()
plt.title('Normal Distribution with 95% Confidence Interval')
plt.xlabel('Revenue')
plt.ylabel('Probability Density')
plt.show()
```

**Q14. A researcher hypothesizes that a new drug will decrease blood pressure by 10 mmHg. They conduct a clinical trial with 100 patients and find that the sample mean decrease in blood pressure is 8 mmHg with a standard deviation of 3 mmHg. Test the hypothesis with a significance level of 0.05.**

To test the hypothesis using a Z-test, you can follow these steps:

$$H_0: \mu = 10$$

2. The alternative hypothesis $H_1$ represents the researcher's claim that the mean decrease in blood pressure $\mu$ is different from 10 mmHg.

$$H_1: \mu \neq 10 $$

3. Set the significance level $\alpha$: The significance level, denoted as $\alpha$, is typically set at 0.05, indicating a 5% significance level.

4. Calculate the test statistic ($Z$): The test statistic ($Z$) for a Z-test is calculated as follows:

$$Z = \frac{\bar{x} - \mu}{\frac{\sigma}{\sqrt{n}}}$$ 

where:

- $\bar{x}$ is the sample mean (8 mmHg)
- $\mu$ is the population mean under the null hypothesis (10 mmHg)
- $\sigma$ is the population standard deviation (3 mmHg)
- $n$ is the sample size (100)

$$Z = \frac{8 - 10}{\frac{3}{\sqrt{100}}} = \frac{-20}{3} = -6.67$$

5. Find the critical value and make a decision: To find the critical value, we can use the Z-table or a Z-distribution calculator. Since we using a two-tailed test. At a 5% significance level, the critical Z-value for a left-tailed test is approximately We use $Z_{\alpha/2} = 1.96$ to find critical values for both tails. Critical value range is $(1.96, -1.96)$.

6. Compare the calculated Z-value with the critical Z-value: If (Z < -1.96), reject the null hypothesis; otherwise, fail to reject it.

7. Make a conclusion: Based on our decision in step 6, making a conclusion regarding the hypothesis test. Therefore, we have sufficient evidence to conclude that the new drug has a statistically significant effect on blood pressure, and the mean decrease in blood pressure $𝜇$ is different from 10 mmHg.

`We can see the below visualization:`

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Define the sample statistics
sample_mean = 8
population_mean = 10
population_stddev = 3
sample_size = 100

# Define the confidence level
confidence_level = 0.95

# Calculate the critical Z-score for a 95% confidence interval
# Set the significance level (alpha) for the test
alpha = 1 - confidence_level
z_alpha_over_2 = norm.ppf(1 - alpha / 2)

# Calculate the margin of error
margin_of_error = z_alpha_over_2 * (population_stddev / np.sqrt(sample_size))

# Calculate the confidence interval
lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

# Define the null and alternative hypotheses
null_hypothesis = "H0: The mean decrease in blood pressure 𝜇 is not different from 10 mmHg."
alternative_hypothesis = "H1: the mean decrease in blood pressure 𝜇 is different from 10 mmHg."

# Calculate the test statistic (Z-score)
z = (sample_mean - population_mean) / (population_stddev / np.sqrt(sample_size))

# Step 7: Print the results
print(f"Test Statistic (Z): {z:.2f}")
print(f"Critical Value (Z_alpha/2): {z_alpha_over_2:.2f}")
print(f"{null_hypothesis}\n{alternative_hypothesis}")
# Perform the hypothesis test
if abs(z) > z_alpha_over_2:
    print(f'Reject the null hypothesis at {confidence_level * 100}% confidence level.')
else:
    print(f'Fail to reject the null hypothesis at {confidence_level * 100}% confidence level.')

# Step 8: Visualize the data
x = np.linspace(-7, 20, 1000)
y = norm.pdf(x, sample_mean, population_stddev)

plt.figure(figsize=(10, 5))
plt.plot(x, y, label='Normal Distribution', color='b')
plt.fill_between(x, 0, y, where=(x >= lower_bound) & (x <= upper_bound), color='yellow', alpha=0.5,
                 label='95% Confidence Interval')
plt.axvline(sample_mean, color='r', linestyle='--', label=f'Sample Mean = {sample_mean:.2f}')
plt.axvline(z, color='brown', linestyle='dashed', linewidth=2, label=f'Z = {z:.2f}')
# Add arrow and text box to display the range values on the x-axis
plt.annotate(f'lower_bound = {lower_bound:.2f}', xy=(lower_bound, 0), xytext=(-5, 0.06),
             arrowprops=dict(arrowstyle='->', color='green'), fontsize=12)
plt.annotate(f'upper_bound = {upper_bound:.2f}', xy=(upper_bound, 0), xytext=(13, 0.06),
             arrowprops=dict(arrowstyle='->', color='green'), fontsize=12)
plt.legend()
plt.title('Normal Distribution with 95% Confidence Interval')
plt.xlabel('Decrease in blood pressure (mmHg)')
plt.ylabel('Probability Density')
plt.show()
```

    - $s$ is the sample standard deviation (0.5 pounds).
    - $n$ is the sample size (25).

$$t = \frac{4.8 - 5}{0.5/\sqrt{25}} = -2$$

   - The rejection region is the range of t-scores that would lead us to reject the null hypothesis. 
   - Since this is a one-tailed test with a left-tailed alternative, we need to find the t-score that has a cumulative probability of 0.01 under the t-distribution with 24 degrees of freedom ($n$ - 1). 
   - Using a t-table or a calculator, we can find that the critical value is approximately -2.492. 
   - The rejection region is then all t-scores less than or equal to -2.492. In other words, we will reject the null hypothesis if $t \leq -2.492$.

- Step 4: `Compare the test statistic to the critical value and make a decision:` 
  - Since our test statistic is -2, which is not less than or equal to -2.492, we do not reject the null hypothesis. 
  - This means that we do not have enough evidence to support the claim that the true mean weight of the products is less than 5 pounds.

- Step 5: `State the conclusion in words:`
  - We can conclude that there is no significant difference between the true mean weight of the products and 5 pounds at a 0.01 level of significance.
  
`We can see the visulazation below:`  

```python
# Import necessary libraries for data analysis and visualization
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Define the null and alternative hypotheses
null_hypothesis = "H0:The true mean weight of the products is equal to 5 pounds"
alternative_hypothesis = "H1: The true mean weight of the products is less than 5 pounds"

# Calculate the test statistic
sample_mean = 4.8  # Sample mean (x̄)
hypothesized_mean = 5  # Hypothesized population mean (μ0)
sample_stddev = 0.5  # Sample standard deviation (s)
sample_size = 25  # Sample size (n)

# Calculate the t-score
t_score = (sample_mean - hypothesized_mean) / (sample_stddev / np.sqrt(sample_size))

# Step 3: Determine the critical value and rejection region
# Significance level
alpha = 0.01

# Degrees of freedom
degrees_of_freedom = sample_size - 1

# Calculate the critical value
critical_value = stats.t.ppf(alpha, df=degrees_of_freedom)

# Step 4: Compare the test statistic to the critical value and make a decision
# Make a decision based on the comparison
if t_score <= critical_value:
    decision = "Reject the null hypothesis"
else:
    decision = "Do not reject the null hypothesis"

# Step 5: State the conclusion in words
if decision == "Reject the null hypothesis":
    conclusion = "There is enough evidence to support the claim that the true mean weight of the products is less than 5 pounds."
else:
    conclusion = "There is no significant difference between the true mean weight of the products and 5 pounds at a 0.01 level of significance."

# Step 6: Visualize the t-distribution and the test statistic
# Generate a range of t-values for the t-distribution
t_values = np.linspace(-4, 10, 1000)

# Calculate the probability density function (PDF) for the t-distribution
pdf = stats.t.pdf(t_values, df=degrees_of_freedom)


# Plot the t-distribution
plt.plot(t_values, pdf, label='t-distribution PDF')

# Highlight the critical region
plt.fill_between(t_values, pdf, where=t_values <= critical_value, alpha=0.4, color='red', label='Rejection Region')

# Mark the test statistic on the plot
plt.axvline(x=t_score, color='green', linestyle='--', label='Test Statistic')

# Mark the sample mean on the plot
plt.axvline(x=sample_mean, color='blue', linestyle='--', label='Sample Mean')

# Set labels and title
plt.xlabel('weight of the products (pounds)')
plt.ylabel('Probability Density')
plt.title('Hypothesis Testing with t-Distribution')

# Add legend
plt.legend()

# Show the plot
plt.show()

# Print the results
print(f"{null_hypothesis}\n{alternative_hypothesis}")
print(f"Test Statistic (t): {t_score:.2f}")
print(f"Critical Value: {critical_value:.2f}")
print("Decision:", decision)
print("Conclusion:", conclusion)
```

**Q16. Two groups of students are given different study materials to prepare for a test. The first group (n1 = 30) has a mean score of 80 with a standard deviation of 10, and the second group (n2 = 40) has a mean score of 75 with a standard deviation of 8. Test the hypothesis that the population means for the two groups are equal with a significance level of 0.01.**

To test the hypothesis that the population means for two groups are equal, we can perform a `two-sample t-test`. 

1. The null and alternative hypotheses can be defined as follows:

   - Null Hypothesis (H0): The population means for the two groups are equal.
   
$$H_0: \mu_1 = \mu_2$$

   - Alternative Hypothesis (H1): The population means for the two groups are not equal. 
   
$$H_1: \mu_1 \neq \mu_2$$

   - Where:

     - $\mu_1$ is the population mean for the first group.
     - $\mu_2$ is the population mean for the second group.
     - The significance level $\alpha$ is given as 0.01.

2. Now, let's calculate the test statistic for the two-sample t-test:

   - The formula for the test statistic for the two-sample t-test is:

$$t = \frac{(\bar{x}_1 - \bar{x}_2)}{\sqrt{\frac{s_1^2}{n_1} + \frac{s_2^2}{n_2}}}$$

   - Where:

     - $\bar{x}_1$ and $\bar{x}_2$ are the sample means for the first and second groups, respectively.
     - $s_1$ and $s_2$ are the sample standard deviations for the first and second groups, respectively.
     - $n_1$ and $n_2$ are the sample sizes for the first and second groups, respectively.
     - Given the values in your question:
       - $\bar{x}_1 = 80$
       - $s_1 = 10$
       - $n_1 = 30$
       - $\bar{x}_2 = 75$
       - $s_2 = 8$
       - $n_2 = 40$
     - Let's calculate the test statistic:

$$t = \frac{(80 - 75)}{\sqrt{\frac{10^2}{30} + \frac{8^2}{40}}}$$

   - The test statistic is:

$$t = \frac{(80 - 75)}{\sqrt{\frac{10^2}{30} + \frac{8^2}{40}}}$$

$$t = \frac{5}{\sqrt{\frac{100}{30} + \frac{64}{40}}}$$

$$t = \frac{5}{\sqrt{3.33 + 1.6}}$$

$$t = \frac{5}{\sqrt{4.93}}$$

$$t = \frac{5}{2.22}$$

$$t = 2.25$$

3. The degrees of freedom are:

$$df = \frac{\left(\frac{10^2}{30} + \frac{8^2}{40}\right)^2}{\frac{\left(\frac{10^2}{30}\right)^2}{29} + \frac{\left(\frac{8^2}{40}\right)^2}{39}}$$

$$df = \frac{(3.33 + 1.6)^2}{\frac{(3.33)^2}{29} + \frac{(1.6)^2}{39}}$$

$$df = \frac{(4.93)^2}{\frac{(11.11)}{29} + \frac{(2.56)}{39}}$$

$$df = \frac{(24.31)}{\frac{(11.11)}{29} + \frac{(2.56)}{39}}$$

$$df = \frac{(24.31)}{0.38 + 0.07}$$

$$df = \frac{(24.31)}{0.45}$$

$$df = 54.02$$

   - Since the degrees of freedom are not a whole number, we can round them down to the nearest integer, which is **54**.

4. Finding the critical value:

   - Now, let's find the critical value from the t-distribution with 54 degrees of freedom and a significance level of 0.01.
   - The critical value is the value of t that corresponds to a probability of 0.005 in the upper tail of the t-distribution, since we are using a two-tailed test.
   - We can use a t-table or a calculator to find the critical value.
   - The critical value is approximately **2.68**.

5. Finally, let's compare the test statistic with the critical value to make a decision:

   - If $|t| > t_{\text{critical}}$, we reject the null hypothesis.

   - In this case, $|t| = |2.25| = 2.25$ and $t_{\text{critical}} = 2.68$.

   - Therefore, $|t| < t_{\text{critical}}$, which means we do not reject the null hypothesis.

6. The conclusion can be stated as follows:

   - There is not enough evidence to reject the null hypothesis that the population means for the two groups are equal at the 0.01 significance level.
   
`We can see the below visualization:`   

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Define the null and alternative hypotheses
null_hypothesis = "H0:The population means for the two groups are equal"
alternative_hypothesis = "H1: The population means for the two groups are not equal"

# Define the parameters for group 1
mean1 = 80
std_dev1 = 10
sample_size1 = 30

# Define the parameters for group 2
mean2 = 75
std_dev2 = 8
sample_size2 = 40

# Set the significance level
alpha = 0.01

# Calculate the test statistic
pool_std_dev = np.sqrt(((std_dev1 ** 2) / sample_size1) + ((std_dev2 ** 2) / sample_size2))
t_statistic = (mean1 - mean2) / pool_std_dev

# Calculate the degrees of freedom
degrees_of_freedom = (pool_std_dev ** 4) / (
    ((std_dev1 ** 2 / sample_size1) ** 2 / (sample_size1 - 1)) +
    ((std_dev2 ** 2 / sample_size2) ** 2 / (sample_size2 - 1))
)

# Calculate the critical value from the t-distribution
t_critical = stats.t.ppf(1 - alpha / 2, degrees_of_freedom)

# Create a range of t values for the plot
t_values = np.linspace(-3, 3, 400)
# Create the t-distribution probability density function
pdf = stats.t.pdf(t_values, degrees_of_freedom)

# Plot the t-distribution
plt.plot(t_values, pdf, label='T-Distribution (df={:.2f})'.format(degrees_of_freedom))

# Fill the critical region
plt.fill_between(t_values, pdf, where=(t_values >= t_critical) | (t_values <= -t_critical), color='red', alpha=0.4, label='Critical Region')
# Show confidence interval
plt.fill_between(t_values, pdf, where=(t_values < t_critical) & (t_values > -t_critical), color='yellow', alpha=0.4, label='Confidence Interval')

# Highlight the test statistic
plt.plot([t_statistic, t_statistic], [0, stats.t.pdf(t_statistic, degrees_of_freedom)], color='green', linestyle='--', label='Test Statistic (t={:.2f})'.format(t_statistic))

# Add labels and legends
plt.xlabel('Population Mean')
plt.ylabel('Probability Density')
plt.title('Two-Sample T-Test')
plt.legend()

# Display the plot
plt.show()

# print the null and alternate hypothesis
print(f"{null_hypothesis}\n{alternative_hypothesis}")
print(f"Test Statistic (t): {t_statistic:.2f}")
print(f"Critical Value: {t_critical:.2f}")
# Make a decision based on the test statistic and critical value
if abs(t_statistic) > t_critical:
    print("Decision: Reject the null hypothesis")
    print("Conclusion: There is enough evidence to reject the null hypothesis that the population means for the two groups are equal at the 0.01 significance level.")
    
else:
    print("Decision: Do not reject the null hypothesis")
    print("Conlusion: There is not enough evidence to reject the null hypothesis that the population means for the two groups are equal at the 0.01 significance level.")
```

**Q17. A marketing company wants to estimate the average number of ads watched by viewers during a TV program. They take a random sample of 50 viewers and find that the sample mean is 4 with a standard deviation of 1.5. Estimate the population mean with a 99% confidence interval.**

We are using a `two-tailed test.` A `two-tailed test` means that we are testing for the possibility of the `population mean being either greater than or less than the sample mean.` Therefore, we need to split the significance level α into two halves, one for each tail of the normal distribution. The critical value $Z_{\alpha/2}$ is the value that cuts off the upper $\alpha/2$ area in the right tail and the lower $\alpha/2$ area in the left tail of the normal distribution. This ensures that the total area in both tails is equal to α, which is the probability of making a type I error (rejecting the null hypothesis when it is true).

To estimate the population mean with a 99% confidence interval, we need to use the following formula:

$$\bar{x} \pm z_{\alpha/2} \frac{\sigma}{\sqrt{n}}$$

where 
- $\bar{x}$ is the sample mean
- $z_{\alpha/2}$ is the critical value of the standard normal distribution for a given significance level $\alpha$
- $\sigma$ is the sample standard deviation
- $n$ is the sample size.

In this case, we have:

- $\bar{x} = 4$
- $z_{\alpha/2} = z_{0.005} = 2.576$ (we can find this value from a Z-table or use a calculator)
- $\sigma = 1.5$
- $n = 50$

Plugging these values into the formula, we get:

$$4 \pm 2.576 \frac{1.5}{\sqrt{50}}$$

Simplifying, we get:

$$4 \pm 0.548$$

Therefore, the 99% confidence interval for the population mean is:

$$(3.452, 4.548)$$

This means that we are 99% confident that the true average number of ads watched by viewers during a TV program is between 3.452 and 4.548.

`We can see the visulaization in below:`

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Define the sample statistics
sample_mean = 4
population_std = 1.5
sample_size = 50
# Define the confidence level
confidence_level = 0.99

# Step 4: Calculate the margin of error (E)
# Calculate the critical Z-score for a 95% confidence interval
alpha = 1 - confidence_level
z_alpha_over_2 = norm.ppf(1 - alpha / 2)

# Calculate the margin of error
margin_of_error = z_alpha_over_2 * (population_std / np.sqrt(sample_size))

# Step 5: Calculate the confidence interval
lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

# Step 7: Print the results
print(f"We are {confidence_level * 100}% confident that the true population mean revenue falls within the interval "
      f"({lower_bound:.2f}, {upper_bound:.2f}).")

# Step 8: Visualize the data
x = np.linspace(sample_mean - 3 * population_std, sample_mean + 3 * population_std, 1000)
y = norm.pdf(x, sample_mean, population_std)

plt.figure(figsize=(10, 5))
plt.plot(x, y, label='Normal Distribution', color='b')
# fill the confidence interval
plt.fill_between(x, 0, y, where=(x >= lower_bound) & (x <= upper_bound), color='yellow', alpha=0.4,
                 label='95% Confidence Interval')
# Fill the critical region
plt.fill_between(x, 0, y, where=(x < lower_bound) | (x > upper_bound), color='red', alpha=0.4, label='Critical Region')
plt.axvline(sample_mean, color='r', linestyle='--', label='Sample Mean')
# Add arrow and text box to display the range values on the x-axis
plt.annotate(f'lower_bound = {lower_bound:.2f}', xy=(lower_bound, 0), xytext=(0, 0.15),
             arrowprops=dict(arrowstyle='->', color='green'), fontsize=12)
plt.annotate(f'upper_bound = {upper_bound:.2f}', xy=(upper_bound, 0), xytext=(6, 0.15),
             arrowprops=dict(arrowstyle='->', color='green'), fontsize=12)
plt.legend()
plt.title('Normal Distribution with 95% Confidence Interval')
plt.xlabel('No of Ads Watched By Viewers During a TV Program')
plt.ylabel('Probability Density')
plt.show()
```
