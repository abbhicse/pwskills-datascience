**Q1. What is Flask Framework? What are the advantages of Flask Framework?**

`Flask` is a `lightweight web framework` written in Python. It allows you to build web applications quickly and easily. `Flask` follows the `WSGI (Web Server Gateway Interface)` standard and is based on the` Werkzeug toolkit` and the `Jinja2 template engine.`

`Advantages of Flask Framework:`

- `Lightweight:` Flask is a minimalistic framework with a small core. It provides only the essentials for web development, allowing for flexibility and simplicity.

- `Easy to Get Started:` Flask has a straightforward and intuitive API that makes it easy to learn and get started with. You can quickly set up a basic web application with just a few lines of code.

- `Flexible:` Flask doesn't enforce any particular way of structuring your application. It gives you the freedom to choose the tools and libraries you want to use, allowing for more flexibility in development.

- `Modular:` Flask is designed as a collection of small and independent modules that can be combined to create a web application. This modular approach makes it easy to add or remove functionality as needed.

- `Extensible:` Flask provides a wide range of extensions that can add additional functionality to your application, such as support for `database integration`, `form validation`, `authentication`, and more.

- `Well-documented:` Flask has comprehensive documentation that covers all aspects of the framework. This makes it easy to find answers to your questions and get help when needed.

Overall, `Flask` is a powerful and versatile framework that provides a solid foundation for building web applications in Python.

**Q2. Create a simple Flask application to display ‘Hello World!!’. Attach the screenshot of the output in Jupyter Notebook.**

```python
from flask import Flask

app = Flask(__name__)

@app.route('/')
def hello_world():
    return 'Hello World!!'

if __name__ == '__main__':
    app.run()
```

`The above code doing these things:`

- Imported the `Flask` module.
- Created a `Flask` application instance.
- Defined a `route` for the root `URL ("/")` and a corresponding function that returns the `"Hello World!!"` string.
- Added a `conditional` statement to ensure that the `development server` is only run when the `script` is executed directly.
- Started the `Flask development server.`

**Q3. What is App routing in Flask? Why do we use app routes?**

`App routing` in Flask refers to the process of `mapping URLs (routes)` to specific functions in a Flask application. It allows you to define different `routes` for different `URLs` and associate them with corresponding functions that handle the requests for those `URLs.`

`App routes` are used in Flask to determine which function should be executed when a `particular URL` is accessed by a `client.` By defining routes, you can create different `endpoints` within your application and define the `logic` that should be executed for `each endpoint`. This enables you to handle different types of `requests` and provide appropriate `responses` based on the requested `URL.`

**Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.**

In `Flask`, the `url_for()` function is used for `URL building`. It generates a `URL` for the specified `endpoint` and allows you to `dynamically` create `URLs` based on the `endpoint's name.`

Here's an example code to demonstrate the working of the `url_for()` function:

```python
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def index():
    return 'This is the index page.'

@app.route('/user/<username>')
def profile(username):
    return f'This is the profile page of {username}.'

@app.route('/login')
def login():
    return 'This is the login page.'

with app.test_request_context():
    print(url_for('index'))  # Output: /
    print(url_for('profile', username='abhishek'))  # Output: /user/John
    print(url_for('login'))  # Output: /login
```

The `url_for()` function takes the `name` of the `endpoint` as its `first argument` and `optional` keyword arguments for any `route parameters`. It returns the `URL associated` with the specified `endpoint`.

In the example code, we have three routes: `'/'` for the `index page`, `'/user/<username>'` for the `profile` page, and `'/login'` for the `login` page. We use the `url_for()` function to generate the `URLs` for these `endpoints.`
