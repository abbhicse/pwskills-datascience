  (ii) Develop a program to compute the mode of a list of integers.
  
  (iii) Implement a function to calculate the weighted mean of a list of values and their corresponding weights.
  
  (iv) Write a Python function to find the geometric mean of a list of positive numbers.

  (vi) Build a function to determine the midrange of a list of numbers (average of the minimum and maximum).
  
  (vii) Implement a Python program to find the trimmed mean of a list, excluding a certain percentage of outliers.

````python
import random

int_list = [random.randint(90, 130) for _ in range(100)]

print(int_list)
````

````python
import statistics

# (i) Calculate mean and median
def calculate_mean(numbers):
    return sum(numbers) / len(numbers)

def calculate_median(numbers):
    return statistics.median(numbers)

# (ii) Compute mode
def compute_mode(numbers):
    return statistics.mode(numbers)

# (iii) Calculate weighted mean
def calculate_weighted_mean(data, weights):
    total_weighted_sum = sum(x * w for x, w in zip(data, weights))
    total_weights = sum(weights)
    return total_weighted_sum / total_weights

# (iv) Find geometric mean
def find_geometric_mean(numbers):
    product = 1
    for num in numbers:
        product *= num
    return product ** (1 / len(numbers))

# (v) Calculate harmonic mean
def calculate_harmonic_mean(numbers):
    return len(numbers) / sum(1 / num for num in numbers)

# (vi) Determine midrange
def determine_midrange(numbers):
    return (min(numbers) + max(numbers)) / 2

# (vii) Find trimmed mean
def find_trimmed_mean(numbers, trim_percent):
    trimmed_count = int(len(numbers) * trim_percent)
    trimmed_numbers = sorted(numbers)[trimmed_count : -trimmed_count]
    return sum(trimmed_numbers) / len(trimmed_numbers)

# Test the functions
print("Mean:", calculate_mean(int_list))
print("Median:", calculate_median(int_list))
print("Mode:", compute_mode(int_list))
print("Weighted Mean:", calculate_weighted_mean(int_list, [1] * len(int_list)))
print("Geometric Mean:", find_geometric_mean(int_list))
print("Harmonic Mean:", calculate_harmonic_mean(int_list))
print("Midrange:", determine_midrange(int_list))
print("Trimmed Mean (10%):", find_trimmed_mean(int_list, 0.1))
````

````python
import statistics

# (ii) Calculate range
def calculate_range(numbers):
    return max(numbers) - min(numbers)

# (iii) Calculate variance and standard deviation
def calculate_variance_and_std_dev(numbers):
    variance = statistics.variance(numbers)
    std_dev = statistics.stdev(numbers)
    return variance, std_dev

# (iv) Compute interquartile range (IQR)
def compute_interquartile_range(numbers):
    quartile = statistics.quantiles(numbers, n=4)
    return quartile[2] - quartile[0]

# (v) Calculate coefficient of variation
def calculate_coefficient_of_variation(numbers):
    mean = statistics.mean(numbers)
    std_dev = statistics.stdev(numbers)
    return (std_dev / mean) * 100

# (vi) Calculate mean absolute deviation (MAD)
def calculate_mean_absolute_deviation(numbers):
    mean = statistics.mean(numbers)
    mad = sum(abs(x - mean) for x in numbers) / len(numbers)
    return mad

# (vii) Calculate quartile deviation
def calculate_quartile_deviation(numbers):
    quartile = statistics.quantiles(numbers, n=4)
    return (quartile[2] - quartile[0]) / 2

# (viii) Calculate range-based coefficient of dispersion
def calculate_range_based_coefficient_of_dispersion(numbers):
    return calculate_range(numbers) / (max(numbers) + min(numbers)) * 100

# Test the functions
print("Range:", calculate_range(int_list2))
variance, std_dev = calculate_variance_and_std_dev(int_list2)
print("Variance:", variance)
print("Standard Deviation:", std_dev)
print("Interquartile Range (IQR):", compute_interquartile_range(int_list2))
print("Coefficient of Variation:", calculate_coefficient_of_variation(int_list2))
print("Mean Absolute Deviation (MAD):", calculate_mean_absolute_deviation(int_list2))
print("Quartile Deviation:", calculate_quartile_deviation(int_list2))
print("Range-Based Coefficient of Dispersion:", calculate_range_based_coefficient_of_dispersion(int_list2))
````

3. Write a Python class representing a discrete random variable with methods to calculate its expected value and variance.

````python
class DiscreteRandomVariable:
    def __init__(self, values, probabilities):
        self.values = values
        self.probabilities = probabilities

    def expected_value(self):
        return sum(x * p for x, p in zip(self.values, self.probabilities))

    def variance(self):
        mean = self.expected_value()
        return sum((x - mean) ** 2 * p for x, p in zip(self.values, self.probabilities))


values = [1, 2, 3, 4, 5, 6]
probabilities = [1/6, 1/6, 1/6, 1/6, 1/6, 1/6]
random_variable = DiscreteRandomVariable(values, probabilities)
print("Expected Value:", random_variable.expected_value())
print("Variance:", random_variable.variance())
````

4. Implement a program to simulate the rolling of a fair six-sided die and calculate the expected value and variance of the outcomes.

````python
import random

def simulate_die_rolling(n_simulations):
    outcomes = [random.randint(1, 6) for _ in range(n_simulations)]
    expected_value = sum(outcomes) / n_simulations
    variance = sum((x - expected_value) ** 2 for x in outcomes) / n_simulations
    return expected_value, variance


n_simulations = 100000
expected_value, variance = simulate_die_rolling(n_simulations)
print("Expected Value (Simulation):", expected_value)
print("Variance (Simulation):", variance)
````

  To generate the data use the following command:
  ```python

  # 50 purchases out of 1000 visitors
  old_layout = np.array([1] * 50 + [0] * 950)
  # 70 purchases out of 1000 visitors  
  new_layout = np.array([1] * 70 + [0] * 930)
  ```

  Apply z-test to find which layout is successfull.

````python
import numpy as np
import statsmodels.stats.weightstats as stats

# 50 purchases out of 1000 visitors
old_layout = np.array([1] * 50 + [0] * 950)
# 70 purchases out of 1000 visitors
new_layout = np.array([1] * 70 + [0] * 930)

z, p = stats.ztest(old_layout, new_layout, abs(old_layout.mean()-new_layout.mean()))
print("Z-score:", z)
print("p-value:", p)

alpha = 0.05
if p < alpha:
    print("Reject null hypothesis: New layout has a significant effect.")
else:
    print("Fail to reject null hypothesis: No significant effect.")
````


  Use z-test to find if the claims made by tutor are true or false.

````python
before_program = np.array([75, 80, 85, 70, 90, 78, 92, 88, 82, 87])
after_program = np.array([80, 85, 90, 80, 92, 80, 95, 90, 85, 88])

import scipy.stats as stats

mean_before = np.mean(before_program)
mean_after = np.mean(after_program)
std_dev = np.std(after_program, ddof=1)

n = len(before_program)
z = (mean_after - mean_before) / (std_dev / np.sqrt(n))

print("Z-score:", z)

alpha = 0.05
critical_value = stats.norm.ppf(1 - alpha/2)

if abs(z) > critical_value:
    print("Reject null hypothesis: Program has a significant effect on scores.")
else:
    print("Fail to reject null hypothesis: No significant effect.")
````

15. A company is testing two different website layouts to see which one leads to higher click-through rates. Write a Python function to perform an A/B test analysis, including calculating the t-statistic, degrees of freedom, and p-value.

  Use the following data:
```python
layout_a_clicks = [28, 32, 33, 29, 31, 34, 30, 35, 36, 37]
layout_b_clicks = [40, 41, 38, 42, 39, 44, 43, 41, 45, 47]
```

````python
layout_a_clicks = [28, 32, 33, 29, 31, 34, 30, 35, 36, 37]
layout_b_clicks = [40, 41, 38, 42, 39, 44, 43, 41, 45, 47]
````

````python
import numpy as np
import scipy.stats as stats

def perform_ab_test(layout_a_data, layout_b_data):
    # Calculate sample statistics
    mean_a = np.mean(layout_a_data)
    mean_b = np.mean(layout_b_data)
    std_dev_a = np.std(layout_a_data, ddof=1)
    std_dev_b = np.std(layout_b_data, ddof=1)
    n_a = len(layout_a_data)
    n_b = len(layout_b_data)

    # Calculate pooled standard deviation
    pooled_std_dev = np.sqrt(((n_a - 1) * std_dev_a**2 + (n_b - 1) * std_dev_b**2) / (n_a + n_b - 2))

    # Calculate t-statistic
    t_statistic = (mean_a - mean_b) / (pooled_std_dev * np.sqrt(1/n_a + 1/n_b))

    # Calculate degrees of freedom
    degrees_of_freedom = n_a + n_b - 2

    # Calculate p-value
    p_value = 2 * (1 - stats.t.cdf(abs(t_statistic), df=degrees_of_freedom))

    return t_statistic, degrees_of_freedom, p_value



t_stat, dof, p_val = perform_ab_test(layout_a_clicks, layout_b_clicks)

print("T-Statistic:", t_stat)
print("Degrees of Freedom:", dof)
print("P-Value:", p_val)

alpha = 0.05
if p_val < alpha:
    print("Reject null hypothesis: There is a statistically significant difference in click-through rates.")
else:
    print("Fail to reject null hypothesis: No statistically significant difference.")
````

18. An HR department wants to investigate if there's a gender-based salary gap within the company. Develop a program to analyze salary data, calculate the t-statistic, and determine if there's a statistically significant difference between the average salaries of male and female employees.

  Use the below code to generate synthetic data:

  ```python
  # Generate synthetic salary data for male and female employees
  np.random.seed(0)  # For reproducibility
  male_salaries = np.random.normal(loc=50000, scale=10000, size=20)
  female_salaries = np.random.normal(loc=55000, scale=9000, size=20)
  ```

````python
import numpy as np
import scipy.stats as stats

# Generate synthetic salary data for male and female employees
np.random.seed(0)  # For reproducibility
male_salaries = np.random.normal(loc=50000, scale=10000, size=20)
female_salaries = np.random.normal(loc=55000, scale=9000, size=20)

def analyze_salary_gap(male_salaries, female_salaries):
    # Calculate sample statistics
    mean_male = np.mean(male_salaries)
    mean_female = np.mean(female_salaries)
    std_dev_male = np.std(male_salaries, ddof=1)
    std_dev_female = np.std(female_salaries, ddof=1)
    n_male = len(male_salaries)
    n_female = len(female_salaries)

    # Calculate pooled standard deviation
    pooled_std_dev = np.sqrt(((n_male - 1) * std_dev_male**2 + (n_female - 1) * std_dev_female**2) / (n_male + n_female - 2))

    # Calculate t-statistic
    t_statistic = (mean_female - mean_male) / (pooled_std_dev * np.sqrt(1/n_male + 1/n_female))

    # Calculate degrees of freedom
    degrees_of_freedom = n_male + n_female - 2

    # Calculate p-value
    p_value = 2 * (1 - stats.t.cdf(abs(t_statistic), df=degrees_of_freedom))

    return t_statistic, degrees_of_freedom, p_value

t_stat, dof, p_val = analyze_salary_gap(male_salaries, female_salaries)

print("T-Statistic:", t_stat)
print("Degrees of Freedom:", dof)
print("P-Value:", p_val)

alpha = 0.05
if p_val < alpha:
    print("Reject null hypothesis: There is a statistically significant salary gap.")
else:
    print("Fail to reject null hypothesis: No statistically significant salary gap.")
````

  Use the below data of scores:
  ```python
  branch_a_scores = [4, 5, 3, 4, 5, 4, 5, 3, 4, 4, 5, 4, 4, 3, 4, 5, 5, 4, 3, 4, 5, 4, 3, 5, 4, 4, 5, 3, 4, 5, 4]
  branch_b_scores = [3, 4, 2, 3, 4, 3, 4, 2, 3, 3, 4, 3, 3, 2, 3, 4, 4, 3, 2, 3, 4, 3, 2, 4, 3, 3, 4, 2, 3, 4, 3]
  ```

````python
import numpy as np
import scipy.stats as stats

def analyze_customer_satisfaction(branch_a_scores, branch_b_scores):
    # Calculate sample statistics
    mean_a = np.mean(branch_a_scores)
    mean_b = np.mean(branch_b_scores)
    std_dev_a = np.std(branch_a_scores, ddof=1)
    std_dev_b = np.std(branch_b_scores, ddof=1)
    n_a = len(branch_a_scores)
    n_b = len(branch_b_scores)

    # Calculate pooled standard deviation
    pooled_std_dev = np.sqrt(((n_a - 1) * std_dev_a**2 + (n_b - 1) * std_dev_b**2) / (n_a + n_b - 2))

    # Calculate t-statistic
    t_statistic = (mean_a - mean_b) / (pooled_std_dev * np.sqrt(1/n_a + 1/n_b))

    # Calculate degrees of freedom
    degrees_of_freedom = n_a + n_b - 2

    # Calculate p-value
    p_value = 2 * (1 - stats.t.cdf(abs(t_statistic), df=degrees_of_freedom))

    return t_statistic, degrees_of_freedom, p_value

branch_a_scores = [4, 5, 3, 4, 5, 4, 5, 3, 4, 4, 5, 4, 4, 3, 4, 5, 5, 4, 3, 4, 5, 4, 3, 5, 4, 4, 5, 3, 4, 5, 4]
branch_b_scores = [3, 4, 2, 3, 4, 3, 4, 2, 3, 3, 4, 3, 3, 2, 3, 4, 4, 3, 2, 3, 4, 3, 2, 4, 3, 3, 4, 2, 3, 4, 3]

t_stat, dof, p_val = analyze_customer_satisfaction(branch_a_scores, branch_b_scores)

print("T-Statistic:", t_stat)
print("Degrees of Freedom:", dof)
print("P-Value:", p_val)

alpha = 0.05
if p_val < alpha:
    print("Reject null hypothesis: There is a statistically significant difference in customer satisfaction.")
else:
    print("Fail to reject null hypothesis: No statistically significant difference.")
````

  ```python
  # Sample data: Customer satisfaction scores for each product version
  standard_scores = [80, 85, 90, 78, 88, 82, 92, 78, 85, 87]
  ```
