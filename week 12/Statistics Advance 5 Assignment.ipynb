**Q1. Calculate the 95% confidence interval for a sample of data with a mean of 50 and a standard deviation of 5 using Python. Interpret the results.**

A `confidence interval` is a range of values that estimates an unknown population parameter, such as a mean, a proportion, or a correlation. A confidence interval has a certain level of confidence, which indicates how often the interval will contain the true parameter value if we repeat the sampling process many times. For example, a 95% confidence interval means that we are 95% confident that the interval covers the true parameter value.

To calculate a confidence interval for a sample mean, we need to know the sample mean, the sample size, the population standard deviation, and the desired confidence level. The formula for the confidence interval is:

$$\bar{x} \pm z \frac{\sigma}{\sqrt{n}}$$

where $\bar{x}$ is the sample mean, $z$ is the critical value from the standard normal distribution that corresponds to the confidence level, $\sigma$ is the population standard deviation, and $n$ is the sample size.

In our case, we have a sample of data with a mean of 50 and a standard deviation of 5. We want to calculate the 95% confidence interval. To do this, we need to find the critical value $z$ for the 95% confidence level. We can use the `search_web` tool to look up the value in a table or use a Python function to calculate it. For example, using `search_web(query="confidence interval table")`, I found this table that shows that $z=1.96$ for the 95% confidence level.

Now, we can plug in the values into the formula and get:

$$50 \pm 1.96 \frac{5}{\sqrt{n}}$$

This is the general expression for the 95% confidence interval for the sample mean. To get a specific interval, you need to know the sample size $n$. For example, if $n=100$, then the interval is:

$$50 \pm 1.96 \frac{5}{\sqrt{100}} = 50 \pm 0.98 = [49.02, 50.98]$$

This means that we are 95% confident that the true population mean is between 49.02 and 50.98.

To interpret this result, we can say that if we repeated this sampling process many times and calculated a 95% confidence interval each time, about 95% of those intervals would contain the true population mean.

`Now we can visualize the answer using the below code:`

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

# Sample data properties
sample_mean = 50
population_stddev = 5
sample_size = 100
confidence_level = 0.95  # 95% confidence level

# Calculate the critical value (z) for the given confidence level
z = norm.ppf((1 + confidence_level) / 2)

# Calculate the margin of error
margin_of_error = z * (population_stddev / np.sqrt(sample_size))

# Calculate the confidence interval bounds
lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

# Plotting the confidence interval
x = np.linspace(sample_mean - 3 * margin_of_error, sample_mean + 3 * margin_of_error, 1000)
y = norm.pdf(x, sample_mean, population_stddev / np.sqrt(sample_size))

plt.figure(figsize=(8, 6))
plt.plot(x, y, label='Probability Density Function', color='b')
plt.fill_between(x, 0, y, where=(x >= lower_bound) & (x <= upper_bound), color='g', alpha=0.5, label='95% Confidence Interval')
plt.axvline(sample_mean, color='r', linestyle='--', label='Sample Mean')
plt.title('Sample Mean with 95% Confidence Interval')
plt.xlabel('Sample Mean')
plt.ylabel('Probability Density')
plt.legend()
plt.grid()
plt.show()

# Print the required information
print(f'Sample Mean: {sample_mean:.2f}')
print(f'Population Standard Deviation: {population_stddev:.2f}')
print(f'Sample Size: {sample_size}')
print(f'Critical Z Score: {z:.2f}')
print(f'95% Confidence Interval: [{lower_bound:.2f}, {upper_bound:.2f}]')
```

**Q2. Conduct a chi-square goodness of fit test to determine if the distribution of colors of M&Ms in a bag matches the expected distribution of 20% blue, 20% orange, 20% green, 10% yellow, 10% red, and 20% brown. Use Python to perform the test with a significance level of 0.05.**

A `chi-square goodness of fit test` is a way to check if a categorical variable follows a certain distribution. For example, we can use it to see if the colors of M&Ms in a bag are as expected or not.

To perform the test, we need to follow these steps:

1. `Define the null and alternative hypotheses:` The null hypothesis is that the observed distribution matches the expected distribution. The alternative hypothesis is that they are different.
2. `Calculate the expected counts for each category based on the expected proportions and the total number of observations:` For example, if we have 100 M&Ms in a bag and we expect 20% of them to be blue, then the expected count for blue is 0.2 x 100 = 20.
3. `Calculate the chi-square test statistic using this formula:`

$$\chi^2 = \sum_{i=1}^{k} \frac{(O_i - E_i)^2}{E_i}$$

where $O_i$ is the observed count for category $i$, $E_i$ is the expected count for category $i$, and $k$ is the number of categories.

4. `Find the p-value that corresponds to the chi-square test statistic and the degrees of freedom:` The degrees of freedom are equal to $k-1$, where $k$ is the number of categories. The p-value is the probability of getting a chi-square test statistic equal to or larger than the observed one, assuming that the null hypothesis is true. You can use a chi-square table or a calculator to find the p-value.
5. `Compare the p-value with the significance level (usually 0.05):` If the p-value is less than the significance level, we can reject the null hypothesis and conclude that the observed distribution is different from the expected distribution. If the p-value is greater than or equal to the significance level, we cannot reject the null hypothesis and conclude that there is not enough evidence to say that the observed distribution is different from the expected distribution.

Here is an example of how to apply these steps using Python:

Suppose we have a bag of 120 M&Ms and we want to test if their colors follow the expected distribution of 20% blue, 20% orange, 20% green, 10% yellow, 10% red, and 20% brown. we count the number of M&Ms for each color and get these results:

| Green | 30             | 0.2 * 120 = 24 |
| Yellow| 12             | 0.1 * 120 = 12 |

Okay, I will calculate the chi square value, the degree of freedom, and the p value for the data you provided. Here are the steps:

1. To calculate the chi square value, we need to plug in the observed and expected counts into the formula:

$$\chi^2 = \sum_{i=1}^{k} \frac{(O_i - E_i)^2}{E_i}$$

where $O_i$ is the observed count for category $i$, $E_i$ is the expected count for category $i$, and $k$ is the number of categories.

In this case, $k=6$, so we need to sum up six terms. For example, for blue M&Ms, you have:

$$\frac{(18-24)^2}{24} = 1.5$$

We need to do the same for the other colors and add them up. The final chi square value is:

$$\chi^2 = \frac{(18-24)^2}{24} + \frac{(22-24)^2}{24} + \frac{(30-24)^2}{24} + \frac{(12-12)^2}{12} + \frac{(16-12)^2}{12} + \frac{(22-24)^2}{24}$$

$$\chi^2 = \frac{36}{24} + \frac{4}{24} + \frac{36}{24} + \frac{0}{12} + \frac{16}{12} + \frac{4}{24}$$

$$\chi^2 = 1.5 + 0.1667 + 1.5 + 0 + 1.3333 + 0.1667$$

$$\chi^2 = 4.6667$$

2. To calculate the degree of freedom, you need to use this formula:

$$df = k-1$$

where $k$ is the number of categories.

In this case, $k=6$, so the degree of freedom is:

$$df = 6-1 = 5$$

3. To calculate the p value, you need to use a chi square table or a calculator to find the probability of getting a chi square value equal to or larger than the observed one, assuming that the null hypothesis is true.

In this case, you can use this [online calculator](https://www.socscistatistics.com/pvalues/chidistribution.aspx) to find the p value. You need to enter the chi square value (4.6667) and the degree of freedom (5) and click on "Calculate". The p value is:

$$p = 0.4578$$

This means that there is a 4.57% chance of getting a chi square value as large or larger than 4.6667 if the null hypothesis is true.

`We can use Python to perform the chi-square goodness of fit test as follows:`

```python
# Import scipy.stats module
from scipy.stats import chisquare

# Define observed and expected counts
observed = [18, 22, 30, 12, 16, 22]
expected = [0.2*120, 0.2*120, 0.2*120, 0.1*120, 0.1*120, 0.2*120]

# Perform chi-square goodness of fit test
chi2_stat, p_value = chisquare(observed, expected)

# Print test results
print("Chi-square statistic:", chi2_stat)
print("P-value:", p_value)
```

The p-value is much larger than the significance level of 0.05, so we cannot reject the null hypothesis and conclude that there is not enough evidence to say that the colors of M&Ms in our bag are different from the expected distribution.

There are two types of chi-square tests:
- the `chi-square goodness of fit` test and the `chi-square test of independence.`
- The chi-square goodness of fit test is used when we have one categorical variable and we want to test whether its frequency distribution is different from our expectations.

2. Next, we need to calculate the chi-square statistic, which is the sum of the squared differences between the observed and expected frequencies divided by the expected frequencies. It is a measure of how much the observed frequencies deviate from the expected frequencies.

`The formula for the chi-square statistic is:`

$$\chi^2 = \sum \frac{(O_{ij} - E_{ij})^2}{E_{ij}}$$

where $O_{ij}$ is the observed frequency of the cell in row $i$ and column $j$, and $E_{ij}$ is the expected frequency of the same cell.

`Using Python, you can perform all these steps using a few lines of code. Here is an example:`

```python
# Import libraries
import numpy as np
from scipy.stats import chi2_contingency

# Create contingency table
observed = np.array([[20, 15], [10, 25], [15, 20]])

# Perform chi-square test
chi2, p, df, expected = chi2_contingency(observed)

# Print results
print("Chi-square statistic:", chi2)
print("P-value:", p)
print("Degrees of freedom:", df)
print("Expected frequencies:")
print(expected)

# Define null and alternate hypothesis
null_hypothesis = "There is no association between the 'group' and 'outcome' variables."
alternate_hypothesis = "There is an association between the 'group' and 'outcome' variables."

# Compare p-value with significance level (e.g., 0.05)
alpha = 0.05
if p <= alpha:
    print("Null hypothesis is rejected.")
    print("Conclusion: " + alternate_hypothesis)
else:
    print("Null hypothesis is not rejected.")
    print("Conclusion: " + null_hypothesis)
```

**Q4. A study of the prevalence of smoking in a population of 500 individuals found that 60 individuals smoked. Use Python to calculate the 95% confidence interval for the true proportion of individuals in the population who smoke.**

A `confidence interval` is a range of values that is likely to contain the true value of a population parameter with a certain level of confidence. For example, a 95% confidence interval means that we are 95% confident that the interval contains the true value of the parameter. A confidence interval can be calculated from a sample statistic and its standard error.

- In this case, the population parameter is `the proportion of individuals who smoke`, denoted by p. The sample statistic is the proportion of individuals who smoke in the sample, denoted by $\hat{p}$. The standard error of $\hat{p}$ is given by the formula:

$$SE(\hat{p}) = \sqrt{\frac{\hat{p}(1-\hat{p})}{n}}$$

- where n is the sample size.

- To calculate the 95% confidence interval for p, we need to find the margin of error, which is the product of the standard error and the critical value. The critical value depends on the level of confidence and the sampling distribution. For a 95% confidence level and a large sample size, we can use the normal distribution and find the critical value from a z-table. The critical value for a 95% confidence level is approximately `1.96.`

- The margin of error is then:

$$ME = 1.96 \times SE(\hat{p})$$

- The 95% confidence interval for p is then:

$$\hat{p} \pm ME$$

- Plugging in the given values, we get:

$$\hat{p} = \frac{60}{500} = 0.12$$

$$SE(\hat{p}) = \sqrt{\frac{0.12(1-0.12)}{500}} = 0.0145$$

$$ME = 1.96 \times 0.0145 = 0.0284$$

$$\hat{p} \pm ME = 0.12 \pm 0.0284 = (0.092, 0.148)$$

- Therefore, the `95% confidence interval` for p is `(0.092, 0.148).` This means that we are 95% confident that the true proportion of individuals who smoke in the population is between `9.2% and 14.8%.`

```python
# Import the libraries
import numpy as np # for numerical calculations
import matplotlib.pyplot as plt # for plotting graphs
import scipy.stats as stats # for statistical functions

# Define the sample size, number of smokers, and proportion of smokers
n = 500 # sample size
x = 60 # number of smokers
p = x / n # proportion of smokers
confidence_level = 0.95  # 95% confidence level

# Calculate the critical value (z) for the given confidence level
z = stats.norm.ppf((1 + confidence_level) / 2)
# Calculate the standard error of the proportion
standard_error = np.sqrt(p * (1 - p) / n)
# Calculate the margin of error
margin_of_error = z * standard_error
# Calculate the 95% confidence interval range
lower_bound = p - margin_of_error # lower bound of the confidence interval
upper_bound = p + margin_of_error # upper bound of the confidence interval

# Print the results
print(f"Critical Z Score: {z:.2f}")
print(f"The proportion of smokers in the population is {p:.3f} with a 95% confidence interval of ({lower_bound:.3f}, {upper_bound:.3f}).")
# Plot the normal distribution with the confidence interval
# Create a range of values from the mean minus 4 standard errors to the mean plus 4 standard errors
x_values = np.linspace(p - 4 * standard_error, p + 4 * standard_error, 100)
# Calculate the probability density function values for each x value
y_values = stats.norm.pdf(x_values, p, standard_error)

plt.figure(figsize=(8, 6))
# Plot the normal curve
plt.plot(x_values, y_values, label='Probability Density Function', color='red')
# Plot the shaded area for the confidence range
plt.fill_between(x_values, y_values, where=(x_values >= lower_bound) & (x_values <= upper_bound), color="green", alpha=0.5, label="95% Confidence Interval")
# Plot the dashed line for the mean
plt.axvline(p, color="blue", linestyle="--", label='Sample Mean')
# Add some labels and title
plt.xlabel("Proportion of smokers")
plt.ylabel("Probability density")
plt.title("95% confidence interval for the proportion of smokers in a population")
plt.legend()
plt.grid()
# Show the plot
plt.show()
```

**Q5. Calculate the 90% confidence interval for a sample of data with a mean of 75 and a standard deviation of 12 using Python. Interpret the results.**

To calculate the 90% confidence interval for a sample of data with a mean of 75 and a standard deviation of 12, we can use the following formula:

$$CI = \bar{X} \pm Z_{\alpha/2} \times \frac{\sigma}{\sqrt{n}}$$

Where:

- $\bar{X}$ is the sample mean
- $Z_{\alpha/2}$ is the Z-value for the given confidence level
- $\sigma$ is the population standard deviation
- $n$ is the sample size

For a 90% confidence level, the Z-value is 1.645. Assuming that the sample size is 100, we can plug in the values and get:

$$CI = 75 \pm 1.645 \times \frac{12}{\sqrt{100}}$$

$$CI = 75 \pm 1.968$$

$$CI = (73.032, 76.968)$$

This means that we are 90% confident that the true population mean is between 73.032 and 76.968.

`To visualize the confidence interval using matplotlib, a Python library for data visualization, we can use the following code:`

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

# Sample data properties
sample_mean = 75
population_stddev = 12
sample_size = 100
confidence_level = 0.9  # 90% confidence level

# Calculate the critical value (z) for the given confidence level
z = norm.ppf((1 + confidence_level) / 2)

# Calculate the margin of error
margin_of_error = z * (population_stddev / np.sqrt(sample_size))

# Calculate the confidence interval bounds
lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

# Plotting the confidence interval
x = np.linspace(sample_mean - 3 * margin_of_error, sample_mean + 3 * margin_of_error, 1000)
y = norm.pdf(x, sample_mean, population_stddev / np.sqrt(sample_size))

plt.figure(figsize=(8, 6))
plt.plot(x, y, label='Probability Density Function', color='b')
plt.fill_between(x, 0, y, where=(x >= lower_bound) & (x <= upper_bound), color='g', alpha=0.5, label='95% Confidence Interval')
plt.axvline(sample_mean, color='r', linestyle='--', label='Sample Mean')
plt.title('Sample Mean with 90% Confidence Interval')
plt.xlabel('Sample Mean')
plt.ylabel('Probability Density')
plt.legend()
plt.grid()
plt.show()

# Print the required information
print(f'Sample Mean: {sample_mean:.2f}')
print(f'Population Standard Deviation: {population_stddev:.2f}')
print(f'Sample Size: {sample_size}')
print(f'Critical Z Score: {z:.2f}')
print(f'95% Confidence Interval: [{lower_bound:.2f}, {upper_bound:.2f}]')
```

First, let me show the code for plotting the chi-square distribution with 10 degrees of freedom using matplotlib, a popular Python library for data visualization. I will also label the axes and shade the area corresponding to a chi-square statistic of 15.

```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2

# Define x-axis range and degrees of freedom
x = np.arange(0, 30, 0.01)
df = 10

# Plot the chi-square distribution curve
plt.plot(x, chi2.pdf(x, df), color='blue', label='Chi-square PDF')
# Plot the dashed line for the mean
plt.axvline(chi2.mean(df), color="green", linestyle="--", label='Chi-square Mean')
# Plot the dashed line for the variance
plt.axvline(chi2.var(df), color="red", linestyle="--", label='Chi-square Variance')
# Label the axes and add a title
plt.xlabel('x')
plt.ylabel('Density')
plt.title('Chi-square Distribution with 10 Degrees of Freedom')

# Shade the area under the curve for x >= 15
plt.fill_between(x, chi2.pdf(x, df), where=(x >= 15), color='red', alpha=0.5, label='Area for x >= 15')
# Add a legend
plt.legend()
# Show the plot
plt.show()
```

Now, let me explain what is a `chi-square distribution` and how it is related to the `standard normal distribution.`

- A `chi-square distribution` is a continuous probability distribution that describes the distribution of the sum of squares of independent standard normal random variables. A `standard normal random` variable is a normal random variable with a mean of zero and a standard deviation of one. It is also called a Z-score.
- For example, if Z1, Z2, ..., Zk are independent standard normal random variables, then the sum of their squares,

$$\chi^2_k = Z_1^2 + Z_2^2 + ... + Z_k^2$$

- is distributed according to the `chi-square distribution` with k degrees of freedom. The parameter k is a positive integer that specifies the number of standard normal random variables being summed. It is also called the `degrees of freedom.`
- The `chi-square distribution` has one parameter: k. The shape of the chi-square distribution depends on the value of k. As k increases, the chi-square distribution becomes more symmetric and bell-shaped. The mean of the chi-square distribution is equal to `k`, and the `variance `is equal to `2k.`
- The chi-square distribution is widely used in statistical hypothesis testing, especially for testing the goodness of fit of a model to a set of data, or testing the independence of two criteria of classification of qualitative data. It is also used for constructing confidence intervals for a population standard deviation of a normal distribution from a sample standard deviation.

**Q7. A random sample of 1000 people was asked if they preferred Coke or Pepsi. Of the sample, 520 preferred Coke. Calculate a 99% confidence interval for the true proportion of people in the population who prefer Coke.**



A `confidence interval` is a range of values that is likely to contain the true value of a population parameter with a certain level of confidence. For example, a 99% confidence interval means that we are 99% confident that the interval contains the true value of the parameter. A confidence interval can be calculated from a sample statistic and its standard error.

- In this case, the population parameter is `the proportion of individuals who  preferred Coke or Pepsi`, denoted by p. The sample statistic is the proportion of individuals who preferred Coke or Pepsi in the sample, denoted by $\hat{p}$. The standard error of $\hat{p}$ is given by the formula:

$$SE(\hat{p}) = \sqrt{\frac{\hat{p}(1-\hat{p})}{n}}$$

- where n is the sample size.

- To calculate the 99% confidence interval for p, we need to find the margin of error, which is the product of the standard error and the critical value. The critical value depends on the level of confidence and the sampling distribution. For a 99% confidence level and a large sample size, we can use the normal distribution and find the critical value from a z-table. The critical value for a 99% confidence level is approximately `2.58.`

- The margin of error is then:

$$ME = 2.58 \times SE(\hat{p})$$

- The 99% confidence interval for p is then:

$$\hat{p} \pm ME$$

- Plugging in the given values, we get:

$$\hat{p} = \frac{520}{1000} = 0.52$$

$$SE(\hat{p}) = \sqrt{\frac{0.52(1-0.52)}{1000}} = 0.0157$$

$$ME = 2.58 \times 0.0157 = 0.0407$$

$$\hat{p} \pm ME = 0.52 \pm 0.0407 = (0.479, 0.561)$$

- Therefore, the `99% confidence interval` for p is `(0.479, 0.561).` This means that we are 99% confident that the true proportion of individuals who  preferred Coke or Pepsi in the population is between `47.9% and 56.1%.`

```python
# Import the libraries
import numpy as np # for numerical calculations
import matplotlib.pyplot as plt # for plotting graphs
import scipy.stats as stats # for statistical functions

n = 1000 # sample size
x = 520 # number of people who preferred Coke or Pepsi
p = x / n # proportion of people who preferred Coke or Pepsi
confidence_level = 0.99  # 99% confidence level

# Calculate the critical value (z) for the given confidence level
z = stats.norm.ppf((1 + confidence_level) / 2)
# Calculate the standard error of the proportion
standard_error = np.sqrt(p * (1 - p) / n)
# Calculate the margin of error
margin_of_error = z * standard_error
# Calculate the 95% confidence interval range
lower_bound = p - margin_of_error # lower bound of the confidence interval
upper_bound = p + margin_of_error # upper bound of the confidence interval

# Print the results
print(f"Critical Z Score: {z:.2f}")
print(f"The proportion of of people who preferred Coke or Pepsi in the population is {p:.3f} with a 99% confidence interval of ({lower_bound:.3f}, {upper_bound:.3f}).")
# Plot the normal distribution with the confidence interval
# Create a range of values from the mean minus 4 standard errors to the mean plus 4 standard errors
x_values = np.linspace(p - 4 * standard_error, p + 4 * standard_error, 100)
# Calculate the probability density function values for each x value
y_values = stats.norm.pdf(x_values, p, standard_error)

plt.figure(figsize=(8, 6))
# Plot the normal curve
plt.plot(x_values, y_values, label='Probability Density Function', color='red')
# Plot the shaded area for the confidence range
plt.fill_between(x_values, y_values, where=(x_values >= lower_bound) & (x_values <= upper_bound), color="green", alpha=0.5, label="95% Confidence Interval")
# Plot the dashed line for the mean
plt.axvline(p, color="blue", linestyle="--", label='Sample Mean')
# Add some labels and title
plt.xlabel("Proportion of people who preferred Coke or Pepsi")
plt.ylabel("Probability density")
plt.title("99% confidence interval for the proportion of people who preferred Coke or Pepsi in a population")
plt.legend()
plt.grid()
# Show the plot
plt.show()
```

**Q8. A researcher hypothesizes that a coin is biased towards tails. They flip the coin 100 times and observe 45 tails. Conduct a chi-square goodness of fit test to determine if the observed frequencies match the expected frequencies of a fair coin. Use a significance level of 0.05.**

A `chi-square goodness of fit test` is a statistical method that compares the observed frequencies of different categories or outcomes with the expected frequencies based on some hypothesis. It can be used to test whether a coin is fair or biased, by comparing the number of heads and tails observed in a series of coin flips with the expected number of heads and tails if the coin was fair.

To conduct a chi-square goodness of fit test, we need to follow these steps:

1. `Define the null and alternative hypotheses:` The null hypothesis is that the coin is fair, meaning that the probability of getting heads or tails is 0.5. The alternative hypothesis is that the coin is biased, meaning that the probability of getting heads or tails is not 0.5.
2. `Calculate the expected frequencies for each category or outcome:` If the coin is fair, then we expect to get 50 heads and 50 tails out of 100 flips. These are the expected frequencies under the null hypothesis.
3. `Calculate the chi-square statistic:` It measures how much the observed frequencies deviate from the expected frequencies. The formula for the chi-square statistic is:

$$\chi^2 = \sum_{i=1}^{k} \frac{(O_i - E_i)^2}{E_i}$$

   - where $k$ is the number of categories or outcomes, $O_i$ is the observed frequency for the $i$-th category or outcome, and $E_i$ is the expected frequency for the $i$-th category or outcome.

   - In this case, we have two categories: heads and tails. The observed frequencies are 55 heads and 45 tails. The expected frequencies are 50 heads and 50 tails. Plugging these values into the formula, we get:

$$\chi^2 = \frac{(55 - 50)^2}{50} + \frac{(45 - 50)^2}{50} = \frac{25}{50} + \frac{25}{50} = 1$$

4. `Find the degrees of freedom:` It is the number of categories or outcomes minus one. In this case, we have two categories: heads and tails. So the degrees of freedom is $2 - 1 = 1$.
5. `Find the p-value:` It is the probability of getting a chi-square statistic as extreme or more extreme than the one we calculated, assuming that the null hypothesis is true. We can use a chi-square distribution table or a calculator to find the p-value. For example, using this [online calculator](https://www.socscistatistics.com/pvalues/chidistribution.aspx), we can enter the chi-square statistic (1) and the degrees of freedom (1) and get the p-value (0.3173).
6. `Compare the p-value with the significance level:` It is the threshold for rejecting or failing to reject the null hypothesis. The significance level is usually denoted by $\alpha$. In this case, $\alpha = 0.05$. If the p-value is less than or equal to $\alpha$, then we reject the null hypothesis and conclude that there is enough evidence to support the alternative hypothesis. If the p-value is greater than $\alpha$, then we fail to reject the null hypothesis and conclude that there is not enough evidence to support the alternative hypothesis.
7. `Interpret the results in terms of the original question:` In this case, since the p-value (0.3173) is greater than $\alpha$ (0.05), we fail to reject the null hypothesis and conclude that there is not enough evidence to support that the coin is biased towards tails.

```python
# Import the libraries
import numpy as np # for numerical calculations
import matplotlib.pyplot as plt # for plotting
from scipy.stats import chi2 # for chi-square distribution

# Define the observed and expected frequencies
observed = np.array([55, 45]) # number of heads and tails observed
expected = np.array([50, 50]) # number of heads and tails expected

# Calculate the chi-square statistic
chi_square = np.sum((observed - expected)**2 / expected) # formula for chi-square statistic

# Find the degrees of freedom
df = len(observed) - 1 # degrees of freedom is number of categories minus one

# Find the p-value
p_value = 1 - chi2.cdf(chi_square, df) # p-value is the area under the curve to the right of the chi-square statistic

# Define the significance level and confidence level
alpha = 0.05 # significance level is the threshold for rejecting or failing to reject the null hypothesis
cl = 1 - alpha # confidence level is the complement of significance level

# Find the critical value and confidence interval
critical_value = chi2.ppf(cl, df) # critical value is the chi-square statistic that corresponds to the confidence level
ci_lower = 0 # lower bound of confidence interval is zero
ci_upper = chi2.ppf(1 - alpha/2, df) # upper bound of confidence interval is the chi-square statistic that corresponds to 1 - alpha/2

# Define the null and alternative hypotheses
null_hypothesis = "The coin is fair (p = 0.5)" # null hypothesis is that the coin is fair, meaning that the probability of getting heads or tails is 0.5
alternative_hypothesis = "The coin is biased (p != 0.5)" # alternative hypothesis is that the coin is biased, meaning that the probability of getting heads or tails is not 0.5

# Draw the conclusion based on the p-value and significance level
if p_value <= alpha: # if p-value is less than or equal to significance level, reject null hypothesis and support alternative hypothesis
    conclusion = "Reject {} and conclude that there is enough evidence to support {}".format(null_hypothesis, alternative_hypothesis)
else: # if p-value is greater than significance level, fail to reject null hypothesis and do not support alternative hypothesis
    conclusion = "Fail to reject {} and conclude that there is not enough evidence to support {}".format(null_hypothesis, alternative_hypothesis)

# Print all the values
print("Null hypothesis:", null_hypothesis) # print null hypothesis
print("Alternative hypothesis:", alternative_hypothesis) # print alternative hypothesis
print("Chi-square statistic:", chi_square) # print chi-square statistic
print("Degrees of freedom:", df) # print degrees of freedom
print("P-value:", p_value) # print p-value
print("\t")
print("Significance level (alpha):", alpha) # print significance level
print("Confidence level (1 - alpha):", cl) # print confidence level
print("Critical value:", critical_value) # print critical value
print("Confidence interval:", (ci_lower, ci_upper)) # print confidence interval
print("Conclusion:", conclusion) # print conclusion

# Plot the chi-square distribution and mark the relevant values
x = np.linspace(0, 10, 100) # create an array of 100 values from 0 to 10
y = chi2.pdf(x, df) # calculate the probability density function of chi-square distribution for each x value
plt.plot(x, y, color="blue", label="Chi-square distribution") # plot the curve
plt.axvline(chi_square, color="red", label="Chi-square statistic") # draw a vertical line at chi-square statistic
plt.axvline(critical_value, color="green", label="Critical value") # draw a vertical line at critical value
plt.fill_between(x[x >= critical_value], y[x >= critical_value], color="orange", label="P-value") # fill the area under the curve to the right of critical value with orange color
plt.fill_between(x[(x >= ci_lower) & (x <= ci_upper)], y[(x >= ci_lower) & (x <= ci_upper)], color="cyan", label="Confidence interval") # fill the area under the curve between lower and upper bounds of confidence interval with cyan color
plt.xlabel("Chi-square") # add x-axis label
plt.ylabel("Probability density") # add y-axis label
plt.title("Chi-square goodness of fit test") # add title
plt.legend() # add legend
plt.show() # show the plot
```

In this case, we want to test if the population mean $\mu$ is significantly different from 70, which is the null hypothesis $H_0: \mu = 70$. The alternative hypothesis $H_a$ is the opposite of the null hypothesis, which is $H_a: \mu \neq 70$.

- To conduct a hypothesis test, we need to calculate a test statistic that measures how far the sample mean $\overline{x}$ is from the hypothesized mean $\mu_0$. Since we do not know the population standard deviation $\sigma$, we use the sample standard deviation $s$ and the t-distribution instead of the z-distribution. The formula for the t-test statistic is:

$$t = \frac{\bar{x} - \mu_0}{s/\sqrt{n}}$$

- where $n$ is the sample size. Plugging in the given values, we get:

$$t = \frac{72 - 70}{10/\sqrt{30}} \approx 1.095$$

- The next step is to find the p-value, which is the probability of getting a test statistic as extreme or more extreme than the observed one, assuming that the null hypothesis is true. Since we have a two-tailed test (because of the $\neq$ sign in $H_a$), we need to find the area under both tails of the t-distribution with $n-1 = 29$ degrees of freedom. We can use a t-table or a calculator to find that:

$$p \approx 0.282528$$

- The final step is to compare the p-value with the significance level $\alpha$, which is the maximum probability of rejecting the null hypothesis when it is true. In this case, $\alpha = 0.05$. If the p-value is less than or equal to $\alpha$, we reject $H_0$ and conclude that there is enough evidence to support $H_a$. If the p-value is greater than $\alpha$, we fail to reject $H_0$ and conclude that there is not enough evidence to support $H_a$.

- Since $p > \alpha$, we fail to reject $H_0$ and conclude that there is not enough evidence to show that the population mean is significantly different from 70.

```python
# Import the libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t

# Define the given values
sample_mean = 72 # The sample mean
sample_std = 10 # The sample standard deviation
sample_size = 30 # The sample size
pop_mean = 70 # The hypothesized population mean
alpha = 0.05 # The significance level

# Calculate the t-test statistic
t_stat = (sample_mean - pop_mean) / (sample_std / np.sqrt(sample_size))
print("The t-test statistic is:", t_stat)

# Calculate the p-value, degree of freedom and significance level
df = sample_size - 1 # The degrees of freedom
p_value = 2 * (1 - t.cdf(abs(t_stat), df)) # The two-tailed p-value
print("The p-value is:", p_value)
print("The degree of freedom is:", df)
print("The significance value is:", alpha)

# Print the hypotheses
print("The null hypothesis is: H0: mu = 70")
print("The alternative hypothesis is: Ha: mu != 70")

# Compare the p-value with the significance level and make a conclusion
if p_value <= alpha:
    print("We reject the null hypothesis and conclude that there is enough evidence to support the alternative hypothesis.")
else:
    print("We fail to reject the null hypothesis and conclude that there is not enough evidence to support the alternative hypothesis.")

# Plot the t-distribution curve
x = np.linspace(65, 75, 100) # The x-axis values
y = t.pdf(x, df) # The y-axis values
plt.plot(x, y, label="t-distribution") # Plot the curve

# Plot the sample mean as a dashed line
plt.axvline(sample_mean, linestyle="--", color="black", label="sample mean")

# Find the critical values and the confidence interval
t_crit = abs(t.ppf(alpha/2, df)) # The critical value for a two-tailed test
lower_bound = pop_mean - t_crit * (sample_std / np.sqrt(sample_size)) # The lower bound of the confidence interval
upper_bound = pop_mean + t_crit * (sample_std / np.sqrt(sample_size)) # The upper bound of the confidence interval

# Shade the rejection region and the acceptance region with different colors
plt.fill_between(x, y, where=(x <= lower_bound) | (x >= upper_bound), color="red", alpha=0.5, label="rejection region")
plt.fill_between(x, y, where=(x > lower_bound) & (x < upper_bound), color="green", alpha=0.5, label="acceptance region")

# Add some labels and a legend
plt.xlabel("t")
plt.ylabel("pdf")
plt.title("t-distribution with sample mean and critical values")
plt.legend()
plt.show()
```
